### 4-1. The Optimisation Problem

#####  

- $d$: 위험 자산 개수  

- $R$: 수익률 벡터 ($N(M, \sum)$)  

- $<x,y>_{\sum^-1} \;= \;x'\sum^{-1}y$  

- $a = <1_d, 1_d>_{\sum^-1}$  

- $b = <M, 1_d>_{\sum^-1}$

- $M - \frac b a 1_d \ne 0$

Var: $\pi' \sum \pi$

return: $\pi'M$

$\pi'1_d = 1$  

따라서 최적화 문제는 아래와 같다.  

$$\text {min}_\pi \pi' \sum \pi = {min}_\pi <\sum x,\sum y>_{\sum^{-1}}$$

$$\pi'M = <\sum \pi, M>_{\sum^{-1}} = m$$

$$\pi'1_d = <\sum \pi, 1_d>_{\sum^{-1}} = 1$$

### 4-2. The Geometric Nature of the Set F(σ,m)

**Corollary**

$\pi_e = \pi_a + \frac{m\pi_e - m\pi_a}{m\omega_{a,b}} \omega_{a,b}$

$(\sigma\pi_e)^2 = (\sigma\pi_a)^2 + \left(\frac{m\pi_e - m\pi_a}{m\omega_{a,b}}\right)^2$

$m\pi_e = m\pi_a + m_{\omega_{a,b}} \sqrt{(\sigma\pi_e)^2 - (\sigma\pi_a)^2} \text{ if } m\pi_e > m\pi_a$

$m\pi_e = m\pi_a - m_{\omega_{a,b}} \sqrt{(\sigma\pi_e)^2 - (\sigma\pi_a)^2} \text{ if } m\pi_e < m\pi_a$

**Efficient and Inefficiend Frontier**  

$F^+ = \pi_a + \lambda\omega_{a,b}, \lambda \geq 0$  
$F^- = \pi_a + \lambda\omega_{a,b}, \lambda \leq 0$  
$F^+(\sigma, m) = (\sigma_\pi, m\_pi ), \pi \in F^+ \text{ and call it the Efficient Frontier,}$  
$F^-(\sigma, m) = (\sigma_\pi, m\_pi ), \pi \in F^- \text{ and call it the Inefficient Frontier.}$

## Chap 6. Mean-Variance Portfolio Theory

### 6-4. Portfolio mean and Variance

$$r = w_1 r_1 + w_2 r_2 + ... + w_n r_n$$  

$$E(r) = w_1 E(r_1) + w_2 E(r_2) + ... + w_n E(r_n)$$  

$$\sigma^2 = \sum_{t=1^n} w_i  w_j  \sigma_{ij}$$

Equal-weighed, 자산간의 상관관계가 없을 때,

$$r = \frac 1 n \sum_{i=1}^n r_i$$  

$$var(r) = \frac 1 {n^2} \sum_{i=1}^n \sigma^2 = \frac {\sigma^2} n$$

상관관계가 있을 때, 

$$var(r) =  \frac {\sigma^2} n + \frac {n - 1} n \bar {Cov}$$

$n$이 커진다면, 결국 공분산만이 남는다. (체계적 위험)

### 6-5. The Feasible Set

- 3개 이상의 자산이 있을 때, Feasible Set은 2차원 영역으로 표시된다.  
- Feasible Set은 왼쪽으로 볼록하다.  

투자자는 당연히 같은 분산이라면 높은 수익을, 같은 수익이라면 낮은 분산을 선호한다. (risk averse)  

따라서 Efficient frontier에만 관심을 갖는다.  

### 6-6. The Markowitz Model

- Nonnegativity Constrains: 만약 $w$를 양수로 제한(공매도 금지)하면 이차계획법(quadratic program) 문제를 풀어야한다.
    - 해당 조건이 없는 경우 최적 투자 비중은 대부분 0이 아닌 값을 갖는다.
    - 해당 조건이 있는 경우 최적 투자 비중은 대부분이 0이다.

### 6-7. The Two-Fund Theorem

Minimum Variance Set의 점 2개에 해당하는 포트폴리오를 알고 있다면, 두 포트폴리오의 조합으로 Minimum Variance Set을 구할 수 있다.  

따라서 모든투자자가 평균과 분산만을 고려하고, 동일한 평균, 분산, 공분산 정보를 이용하며 한 기간 동안 투자한다는 가정이라면, 

두 뮤추얼 펀드의 조합으로 효율적 포트폴리오에 투자할 수 있다.

### 6-8. Inclusion of a Risk-Free Asset

무위험 자산은 분산이 없고 수익만을 생성하므로 Efficient Frontier의 모습은 단순해진다.  

포트폴리오
- 평균 수익률:  $\alpha r_f + (1 - \alpha) \bar r$

### Problem 1.

(Three uncorrelated assets) Suppose there are three uncorrelated assets.  
Each as variance 1, and the mean values are 0.1, 0.2, 0.3 respectively.

**(a)** Find the minimum-variance point.

$$\sigma_p^2 = w_A^2 \sigma_A^2 + w_B^2 \sigma_B^2 + w_C^2 \sigma_C^2 \quad \text {(uncorrelated)}$$

$$= w_A^2 + w_B^2 + w_C^2 \ge \frac {(w_A + w_B + w_C)^2} 3$$

$$(\text {Cauchy–Schwarz inequality}: (a^2 + b^2 + c^2)(x^2 + y^2 + z^2) \ge (ax+by+cz)^2 )$$

$$\sigma_p^2 = 1/3, \quad \sigma = 0.5774, \quad r=0.2 \;\; (w_A=w_B=w_C=\frac 1 3)$$

**(b)**	Draw the feasible region.

```python
import numpy as np
import pandas as pd

asset_rtn = np.array([0.1, 0.2, 0.3])
asset_var = np.array([1,1,1])

port_rtn = []
port_vol = []
port_weights = []

for _ in range(50000):
    weights = np.random.random(asset_rtn.shape[0])
    weights /= np.sum(weights)
    
    rtn = np.sum(asset_rtn * weights)
    var_ = np.sum((weights**2) * (asset_var**2))
    vol = np.sqrt(var_)
    
    port_rtn.append(rtn)
    port_vol.append(vol)
    port_weights.append(weights)
    
pf = pd.DataFrame({'Return': port_rtn, 'Volatility': port_vol, 'Weight': port_weights})
pf.plot(kind='scatter', figsize=(10,6), x='Volatility', y='Return', title='Volatility - Return');
```

```python
mvp_idx = pf['Volatility'].idxmin()
print(f"Volatility: {pf.loc[mvp_idx,'Volatility']:.4f}")
print(f"Return: {pf.loc[mvp_idx,'Return']:.4f}")
print(f"Weight: {np.round(pf.loc[mvp_idx,'Weight'],4)}")
```

### Problem 2.

(Markowitz fun) There are just three assets with rates of return $r_1$, $r_2$ and $r_3$ respectively.  
The covariance matrix and the expected rates of return are 


$$V = \begin{bmatrix}
2 & 1 & 0 \\
1 & 2 & 1 \\
0 & 1 & 2 \\
\end{bmatrix}, \quad \bar r = \begin{bmatrix}
4 \\
8 \\
8 \\
\end{bmatrix}$$

**(a)** Find the minimum-variance portfolio. [Hint: By symmetry $w_1$  = $w_3$]

```python
import numpy as np

cov_ = np.array([
    [2, 1, 0],
    [1, 2, 1],
    [0, 1, 2]
])
inv_cov = np.linalg.inv(cov_)
v1 = np.ones(cov_.shape[0]) @ inv_cov
w1 = v1 / np.sum(v1)

print(f"Weight: {np.round(w1,2)}")
```

**(b)** Find another efficient portfolio by setting $\lambda =1,\mu =0$

```python
r = np.array([0.4, 0.8, 0.8])
v2 = r @ inv_cov
w2 = v2 / np.sum(v2)

print(f"Weight: {np.round(w2,2)}")
```

**(c)** If the risk-free rate is $r_f=0.2$, Find the efficient portfolio of risky assets.

```python
excess_rtn = r - 0.2
v = excess_rtn @ inv_cov
w = v / np.sum(v)

print(f"Weight: {np.round(w,2)}")
```

### Problem 4.

(Two similar assets) Two assets with expected rates of return $r ̅_1$ and $r ̅_2$  have identical
variances and a known correlation coefficient $ρ$.  
There is a risk-free asset with rate of return $r_f$.

**(a)** Find an expression for the optimal (Markowitz) weights for the two assets.

$$
\begin{equation}
    \begin{pmatrix} 
    \sigma_{11} & \sigma_{12}\\
    \sigma_{21} & \sigma_{22}\\
    \end{pmatrix}
    \begin{pmatrix} 
    v_1 \\
    v_2 \\

    \bar r_2 \\
    \end{pmatrix} - r_f
    \begin{pmatrix}

    \end{pmatrix}    
\end{equation}
$$

$$ \sigma_1 = \sigma_2 = \sigma \;\text {라 할 때,} \quad
\begin{equation}
    \begin{pmatrix} 
    \sigma_{11} & \sigma_{12}\\
    \sigma_{21} & \sigma_{22}\\
    \end{pmatrix} = 
    \begin{pmatrix} 
    \sigma^2 & \rho \sigma^2\\
    \rho \sigma^2 & \sigma^2\\
    \end{pmatrix}   
\end{equation}
$$

$$
\begin{equation}
    \begin{pmatrix} 
    v_1 \\
    v_2 \\
    \end{pmatrix} = 
    \begin{pmatrix} 
    \sigma^2 & \rho \sigma^2\\
    \rho \sigma^2 & \sigma^2\\
    \end{pmatrix}  ^{-1} \{
    \begin{pmatrix} 
    \bar r_1 \\
    \bar r_2 \\
    \end{pmatrix} - r_f
    \begin{pmatrix} 
    1 \\
    1 \\
    \end{pmatrix} \}   
\end{equation}
$$

$$
\begin{equation}
    = \frac {\sigma^2} {1-\rho^2}
    \begin{pmatrix} 
    1 & -\rho\\
    -\rho & 1\\
    \end{pmatrix} \{
    \begin{pmatrix} 
    \bar r_1 \\
    \bar r_2 \\
    \end{pmatrix} - r_f
    \begin{pmatrix} 
    1 \\
    1 \\
    \end{pmatrix} \}

$$
\begin{equation}
    = \frac {\sigma^2} {1-\rho^2}

    \bar r_2 - \rho \bar r_1 - r_f + \rho r_f \\

$$v_1 = \frac {\sigma^2} {1-\rho^2} (\bar r_1 - \rho \bar r_2 - r_f + \rho r_f)$$

$$v_2 = \frac {\sigma^2} {1-\rho^2} \bar r_2 - \rho \bar r_1 - r_f + \rho r_f)$$

$$w_1 = \frac {v_1} {v_1 + v_2}$$  

$$w_2 = \frac {v_2} {v_1 + v_2}$$

**(b)** For the parameters $\bar r_1=0.1,\;\; \bar r_2=0.08, \;\;\bar r_f=0.05, \;\;\rho = 0.6$. Find the weight of asset 1. 

```python
r1 = 0.1
r2 = 0.08
rf = 0.05
rho = 0.6

v1 = r1 - rho*r2 - rf + rho*rf
v2 = r2 - rho*r1 - rf + rho*rf

w1 = v1 / (v1 + v2)
w1
```
