### 2.1 Utility Functions and Preferred Investments

**Utility Functions**  

$u: R → R$ such that $u$ is continuous, strictly increasing and twice differentiable. (이 정의에 의해 효용함수는 모두 가역)  

**Preferred Investment, Certain Equivalent and Risk Premium**  

X와 Y가 random payoff일 때
- $X > Y$ (X is preferred to Y for u)$⇔ E(u(X)) ≥ E(u(Y))$  
- $X ≡ Y$ (X is equivalent to Y for u)$⇔ E(u(X)) = E(u(Y))$  

$C_u(X)$가 $u(C_u(X)) = E(u(X))$로부터 정의된 상수일 때,
- $C_u(X)$는 $X$에 대한 certain equivalent
- $\prod_u(X) = E(X) − C_u(X)$는 $X$에 대한 risk premium

이에 대해
- risk taker: $E(X)$보다 $X$를 받는 것을 선호 하는 사람
- risk adverse person: $X$보다 $E(X)$를 받는 것을 선호 하는 사람
- risk neutral person: 차이가 없다고 생각하는 사람  


- for a risk taker $X > E(X)$, i.e. $E(u_r(X)) ≥ u_r(E(X))$,
- for a risk adverse person $E(X) > X$, i.e. $E(u_a(X)) ≤ u_a(E(X))$,
- for a risk neutral person $X ≡ E(X)$, i.e. $E(u_n(X)) = u_n(E(X))$.  

이에 따르면 투자자가 risk adverse일 때 risk premium은 양수이다. (기대 payout을 할인된 가격에 구입함으로써 감수한 위험을 보상)

risk premium이 수학적으로 utility function의 concavity와 연관 어떻게 있는지 설명  

- risk taker: convex $u$
- risk adverse person: concave $u$
- risk neutral person: affine $u$

### 2.2 Gaussian Laws and Mean-Variance Implications

$u$: utility function  
$W_0$: an investment

$$\text {maximize}\; \; E(u(W_T(\pi)))$$

$$W_T(\pi) = W_0(1+R_T(\pi)),\;\; R_T(\pi): \text {strategy} \;\pi \;\text {return}$$

$R_T(\pi) ∼ N(m_\pi, \sigma^2_\pi)$라 하면

$$E(u(W_T(\pi))) = E(u(W_0(1+R_T(\pi))) = E(u(W_0 + m_\pi W_0 + \sigma_\pi W_0Z), \quad Z ∼ N(0,1)$$

$$U(m,\sigma) =  E(u(W_0 + m W_0 + \sigma W_0Z)$$

$$max_\pi E(u(W_T(\pi))) = max_\pi U(m_\pi, \sigma_\pi)$$

### 2.3 Efficient Investment Strategies

### 11.3 Risk Aversion

효용 함수의 주요 목적은 위험 회피 원칙을 포착하여 대안의 순위를 매기는 체계적인 방법을 제공하는 것이다.  
이는 효용 함수가 concave 할 때마다 달성된다.  

[a, b] 범위에서 정의된 $U$가 $0 \le \alpha \le 1$이고, 아래 조건을 만족하면 그 함수는 concave하다.

$$U[\alpha x + (1- \alpha)y] \ge \alpha U(x) + (1 - \alpha)U(y)$$

그리고 concave하다면 risk averse하다고 할 수 있다. (concave: 값이 증가함에 따라 기울기가 평평해짐.(줄어듬))

만약 동전을 던져 앞이 나오면 \\$10을 주고, 뒤가 나오면 주지 않는 게임을 한다고 하고, 돈에 대한 효용함수는 $x-0.04x^2$이라 하자.  
이러한 효용함수를 가진 사람은 \\$5를 확정적으로 주는 것과 이 게임 사이에서 어떤 것을 선택할까?
- 게임을 하는 경우 효용은 3이다. 확정적으로 돈을 받는 경우, 효용은 4이다.
- 따라서 이런 효용함수를 가지고 있다면 확정적을 돈을 받는 것이 이득이다.
- 어디서 바뀔까? 확정적으로 $3.49를 준다고 할 때, 효용은 3이다. 이때 차이가 없어진다.

이는 미분으로도 생각해볼 수 있다.  
- $U'(x)>0$이면 $U(x)$는 strictly increasing이다.
- $U''(x)<0$이면 $U(x)$는 concave하다.

효용함수가 나타내는 위험 회피 정도는 함수의 굴곡(bend)와 관련이 있으며, 커질수록 위험회피도 커진다.  
이는 효용함수의 2차 미분으로 정량화할 수 있다.  

**Arrow-Pratt absolute risk aversion coefficient** (위험 회피 정도)

$$\alpha(x) = - \frac {U''(x)} {U'(x)}$$

$\alpha$는 자산 수준에 따라 위험 회피가 어떻게 변하는지를 나타낸다.  

대부분의 개인은 재산이 증가함에 따라 위험 회피가 감소 -> 재정적으로 안전할 때 더 많은 위험을 감수할 의향이 있다는 사실을 반영  

효용함수 $U(x) = -e^{-ax}$의 경우, $U(x) = ae^{-ax}$ 나 $U(x) = -a^2e^{-ax}$ 나 모두 $\alpha (x) = a$이다.  
이는 모든 $x$에 대해 $\alpha$가 일정한 상수임을 나타낸다. (cf. $U(x) = \text {ln} x$)


**Certainty Equivalent**
기대 효용의 실제 값은 다른 대안과 비교할 때를 제외하고 의미가 없지만, 직관적인 의미를 갖는 측정값이 존재한다.  
x의 Certainty Equivalent는 x의 기대 효용과 동일한 효용 수준을 갖는 특정 양(무위험)으로 정의한다.  

즉, Certainty Equivalent $C$는 다음을 만족한다.  
$$U(C) = E[U(x)]$$

동전 던지기 게임의 Certainty Equivalent는 3.49였는데, 이는 확실히 얻을 수 있는 보상과 동일한 효용을 갖는 값이다.  
효용함수가 concave할 경우, Certainty Equivalent는 항상 기대값보다 작거나 같다. ($C \le E(x)$)

### 11.5 Utility Functions and the Mean-Variance Criterion*

마코위츠 모형에 사용되는 criterion은 
- quadratic utility function
- 수익률의 확률 변수가 가우스 변수라고 가정  

으로서 기대 효용 접근법으로 재구성할 수 있다.  


$$\text {quadratic utility function}: U(x) = ax - \frac 1 2 bx^2, \;\; a > 0, b \ge 0$$

효용함수는 $x < a/b$에서만 증가하기 때문에 해당 범위에서만 유의미하다.

임의의 자산 가치를 y라 할 때,

$$E[U(y)] = E(ay- \frac 1 2 by^2)
= a E(y) - \frac 1 2bE(y^2 )
= aE(y) - \frac 1 2 b[E(y)]^2 - \frac 1 2b\text {var}(y)$$

최적의 포트폴리오는 random variable $y$의 모든 경우에 대해 이 값을 최대화하는 포트폴리오이다.  

- 초기 자산을 1이라 가정하면 $y$는 $R$과 일치한다.  
- $E(y)=M$이라 가정하면, $y$에 대해 최소 분산을 가져야한다. ($E(y)=M=1+m$)
- $y=R$이므로 해는 mean-variance efficient point와 일치해야한다.

다른 mean-variance efficient point는 $a$나 $b$를 다른 값으로 설정함으로써 얻어진다.  
마찬가지로 초기 자신이 1이 아니라면 다른 요인이 소개된다.

### Problem 1. Certainty equivalent

An investor has utility function $U(x)= x^{\frac 1 4}$ for salary.  

He has a new job offer which pays \\$80,000 with a bonus.  

The bonus will be \\$0, \\$10,000, \\$20,000, \\$30,000, \\$40,000, \\$50,000, or \\$60,000, each with equal probability.  

What is the certainty equivalent of this job offer? 

```python
pay = 80000
bonus = [ pay + b * 10000 for b in range(0,7)]
prob = 1 / len(bonus)

utility = list(map(lambda x: prob*(x**(1/4)), bonus))
utility = sum(utility)
fix_pay = utility**4

print(f'Certainty Equivalent: ${int(fix_pay)}')
```

### Problem 2. Wealth independence

Suppose an investor has exponential utility function  $U(x)= -e^{-ax}$ and an initial wealth level of $W$.  

The investor is faced with an opportunity to invest an amount $w ≤W$ and obtain a random payoff $x$.  

Show that his evaluation of this incremental investment is independent of $W$.

$W$의 자산에서 $w$를 투자해 $x$를 얻을 수 있을 때, 전체 자산에 대한 효용함수는 아래와 같이 표현할 수 있다.  

$$E(U(W-w+x)) = E(-e^{-a(W-w+x)}) = E(-e^{-a(W-w)} \times -e^{-ax}) = -e^{-a(W-w)}E(-e^{-ax})$$   

따라서 평가를 하는 것에 있어 $W$는 고려대상이 되지 않는다.

### Problem 4. Relative risk aversion

The Arrow-Pratt relative risk aversion coefficient is

$$μ(x)=  \frac {xU''(x)} {U'(x)}$$

Show that the utility functions $U(x)=\text {ln⁡}x$ and $U(x)= \gamma x^{\gamma}$ have constant relative risk aversion coefficients. 

$$U(x)=\text {ln⁡}x: \quadμ(x)=  \frac {xU''(x)} {U'(x)} = \displaystyle \frac {- \frac {x} {x^2}} {\frac 1 x} = -1$$

$$U(x)= \gamma x^{\gamma}: \quad μ(x)=  \frac {xU''(x)} {U'(x)} = \frac {x \gamma^2 (\gamma-1) x^{(\gamma-2)}} {\gamma^2x^{(\gamma-1)}} = \gamma - 1$$

두 경우 모두 risk aversion coefficient가 상수이므로 $x$에 따라 변하지 않고 일정하다.

### Problem 5. Quadratic mean-variance

An investor with unit wealth maximizes the expected value of the utility function $U(x)= ax- bx^2/2$ and obtains a mean-variance efficient portfolio.  

A friend of his with wealth $W$ and the same utility function does the same calculation, but gets a different portfolio return.  

However, changing $b$ to $b^1$ does yield the same result.  

What is the value of $b'$?

자산이 1일 때, $y = R$이라고 볼 수 있다.  

이에 대해 수익은 $E[U(y)] = 1 \times E[aR-\frac 1 2 b R^2]$이다.  

자산이 $W$일 때, $y = WR$이라고 할 수 있다.  

마찬가지로 아래와 같이 계산할 수 있다.  

$$E[U(y)] = E[aWR-\frac 1 2 b (WR)^2] = W E[aR-\frac 1 2 b WR^2] \quad (aR-\frac 1 2 b WR^2 \neq aR-\frac 1 2 b R^2)$$  

따라서 같은 결과를 얻으려면 $b' = \frac b W$이 되어야한다.
