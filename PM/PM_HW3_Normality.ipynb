### 1.1 Measure of the Perfomance

##### 1.1.1 Return

관측 시점 0, T가 있다고 할 때, 투자 결정이 0 시점에 이루어지며 투자 결과는 T 시점에 관측된다.  

$$ \text {return of the an investment (Price Return)}: R_T = \frac {W_T} {W_0} - 1 $$

$$ \text {return of an asset (Total Return)}: R_T = \frac {W_T \; + \; CR(0, \; T) } {W_0} - 1  $$


$CR(0, \; T)$는 투자기간 중 들어오는 배당이나 쿠폰 등의 수익  


따라서 수익률을 비교할 때는 Total Return을 비교하는 것이 중요
- S&P500 지수는 구성 종목의 배당금을 계산에 포함하지 않음 (Price Return)
- 독일 DAX 30 지수는 구성 종목의 배당금을 포함 (Total Return)

##### 1.1.2 Rate of Return

rate of return(수익률)의 정의는 금융상품이나 시장에 따라 달라질 수 있다.

- monetary rate: 만기 1년 미만인 고정수익상품 (Treasury bill 등)

    $1 + r \; \times T = 1 + R_T $
    
- actuarial rate: 만기 1년 이후인 고정수익상품 (Treasury Note 등) 

    $(1 + r)^T = 1 + R_T$

- continuous rate: Financial Mathematical Modeling

    $ \text {exp} (r \times T) = 1 + R_T $
    
continuous rate(exponential rate)는 아래 성질로 인하여 쉽게 복소화 가능하다.

$$e^{r(0, \; T1 \; + \; T2)(T1\; +\; T2)} = e^{r(0, \;T1)T1} e^{r(T1, \;T1 \;+\; T2)T2}$$

$$\Rightarrow r(0,\; T1 + T2) = \frac {r(0, \;T1)T1 + r(T1, \;T1 + T2)T2} {(T1 + T2)}$$

이는 확률론적 모델링에서 서로 다른 기간의 수익률이 독립 정규 법칙을 따른다고 가정하면, 모든 기간의 수익률도 정규 법칙을 따른다는 것을 의미  

두 기간 사이를 측정하는 방법도 여러가지가 있으므로 사전에 특정 상품에 쓰이는 방식을 숙지할 필요가 있다.

### 1.2 Probabilistic and Empirical Definitions

random variable에 대한 일부 정규성 테스트는 moments를 분석하는 것으로 가능하다.  

random variable의 평균과 표준편차를 이용해 정규화하고, 3차와 4차 moment를 계산하여 수행  

아래에 주어진 모멘트의 확률적 정의는 표본의 실증적 확률에 대한 모멘트를 계산하여 표본에 적용  

**Definition 1.2.1 Probabilistic Definitions for a Random Variable X**

$$\text {Expectation}: E(X) $$  
$$\text {Variance}: Var(X) = E(X^2) - E(X)^2$$  
$$\text {Standard Deviation} : \sigma(X) = \sqrt {Var(X)}$$  
$$\text {Skew}: Skew(X) = E((\frac {X - E(X)} {\sigma (X)})^3)$$  
$$\text {Kurtosis}: Kur(X) = E((\frac {X - E(X)} {\sigma (X)})^4)$$  

**Definition 1.2.2 (Empirical Definitions for a Sample $x = (x_1, x_2,··· x_n)$)**

$$ \text {Sample Mean}: \hat E (x) = \frac{1}{n} \sum_{i=1}^{i=n} x_i,\;\; \text {also denoted} \;\bar x$$
$$ \text {Sample Variance}: \hat {Var}(x) = \frac{1}{n} \sum_{i=1}^{i=n} (x_i - \bar{x})^2$$
$$ \text {Sample Standard Deviation}: \hat {\sigma}(x) = \sqrt{\hat {Var}(x)}$$
$$ \text {Sample Skew}: \hat {Skew}(x) = \frac{1}{n} \sum_{i=1}^{i=n} \left( \frac{x_i - \bar{x}}{\hat \sigma(x)} \right)^3$$
$$ \text {Sample Kurtosis}: \hat {Kurt}(x) = \frac{1}{n} \sum_{i=1}^{i=n} \left( \frac{x_i - \bar{x}}{\hat \sigma(x)} \right)^4$$

Kurtosis가 3을 넘으면 leptokurtic이라 부르며 fat tail이다. (cf. 3보다 낮으면 platykurtic)


### 1.3 Goodness of Fit Tests

Gaussian variable이 Skew(X) = 0, Kurt(X) = 3 이라는 사실로부터, 분포가 가우시안이면 샘플이 많아질수록 Skew와 Kurt는 이 값들로 수렴할 것이다.

**Bera-Jarque theorem**

- 분포가 가우시안이면, 실제로 수렴 값이 존재하며 수렴 속도는 $\sqrt n$이다.
- limit normal distribution은 $\chi^2(2)$분포를 생성하기 위해 더하거나 제곱할 수 있다.
- 따라서 Bera-Jarque Test는 계산된 Skew와 Kurt를 이용해 가설의 강도를 하나의 값으로 나타낼 수 있다.

**Bera-Jarque Test**

$$\hat {BJ} (X) = \frac {n} {6} (\hat {Skew}(X))^2 + \frac {n} {24} (\hat {Kurt}(X) - 3)^2$$  

$$ \hat {BJ} (X) => \chi^2(2)$$

$$ \text {If} \;\; \hat {BJ} (X) > \chi^2_{1-\alpha}(2),\; \text {Bera-Jarque test rejects.}\; (\alpha : \text {confidence level})$$

2017년 DAX30 daily total return을 확인하면 
- [-1.83%, 3.37%] 범위
- 평균: 0.045%
- 표준편차: 0.667% (volatility에 $\sqrt {\Delta T}$를 곱한 값)
- Skew: 0.55
- Kurt: 5.54
- $BJ(X)$: 80.12 
- $\chi^2_{5 \text %}(2)$: 5.99 

$BJ(X)$가 $\chi^2_{5 \text %}(2)$보다 크므로 정규성 가설을 기각한다.  

Skew는 0이 아니므로 비대칭이고, Kurt가 3 이상이므로 fat tail 형태를 갖는 것은 우리가 알고 있는 수렴값과 차이가 있는 것으로 보인다.  

BJ test는 이상치에 민감하다.
- 3.37% 급등이 없던 2017년 2분기에 대해 수행하면 테스트를 통과한다.

### 1.4 Further Statistical Results

정규성 가정을 테스트하는 다른 방법은 밀도 함수 $f$를 추정하기 위해 추정 밀도 함수 $f_n$을 추정하는 것이다.

**Convergence of the Density Function Estimate**: Parzen and Rosenblatt Estimation of the Density

$X$가 밀도함수 $f(x)$를 가지는 random variable일 때, $X_i$는 $X$를 따르는 i.i.d라 하자.

그리고 $K$를 적분해서 1이 나오는 positive 함수라 하고 $h_n$은 0에 수렴하고, $nh_n$은 무한대로 발산한다고 하자.

- 커널함수: https://darkpgmr.tistory.com/147

이 때의 추정 밀도 함수

 

$$
f_n(x) = \frac 1 n \sum_{i=1}^n \frac 1 h_n K (\frac {X_i-x} {h_n})
$$

이산적인 데이터를 이용해 연속 밀도함수를 만드는 것 → 각 값의 커널함수를 더하여 밀도함수를 만듬

- 추정 밀도를 그래프로 그려 정규 분포와 비교하는 시각적 테스트 수행 가능

## Assignment 2. Implementing in Python

The given CSV file contains adjusted closing price data for the top 10 market capitalization stocks on the NASDAQ.  

The data ranges from January 2, 2020, to June 30, 2023.

**(a) Calculate Daily Returns of each Stocks.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_path = 'data/'
nasdaq10 = pd.read_csv(data_path + 'nasdaq10.csv', index_col=0)
nasdaq10
```

```python
rtn_1d = nasdaq10.pct_change(1, axis=0)
rtn_1d = rtn_1d.iloc[1:,:]
rtn_1d
```

```python
cum_rtn_1d = (rtn_1d + 1).cumprod() - 1
cum_rtn_1d.plot(figsize=(15,5), title='Cumulative Return', grid=True, legend=True);
```

**(b) Calculate 1) Min 2) Max 3) Mean 4) Variance 5) Skewness 6) Kurtosis of each Stocks.**

```python
stat_dict = {}
stat_dict['Min'] = rtn_1d.min()
stat_dict['Max'] = rtn_1d.max()
stat_dict['Mean'] = rtn_1d.mean()
stat_dict['Variance'] = rtn_1d.var()
stat_dict['Skewness'] = rtn_1d.skew()
stat_dict['Kurtosis'] = rtn_1d.kurt()

stat_df = pd.DataFrame(stat_dict).transpose()
stat_df
```

**(c) Try Normalization and Draw Normal Distribution Graph of each Stocks.**

```python
normalize_rtn_1d = rtn_1d.apply(lambda x: (x - x.mean()) / x.std())
normalize_rtn_1d
```

```python
from scipy.stats import norm

plt.figure(figsize=(10,6))
plt.title('Distribution')
plt.xlabel('return')
plt.ylabel('pdf')
plt.grid()

for c in rtn_1d.columns:
    sort_x = rtn_1d[c].sort_values()
    plt.plot(sort_x, norm.pdf(sort_x, loc = stat_df.loc['Mean',c], scale = np.sqrt(stat_df.loc['Variance',c])), label = f'{c}')

plt.legend()
plt.show()
```

**(d) Draw QQ-Plots of 10 Stocks and Assess Normality of Returns of each Stocks.**

```python
from scipy.stats import probplot

plt.figure(figsize=(28,10))

col = normalize_rtn_1d.columns
for i in range(len(col)):
    data = normalize_rtn_1d[col[i]]
    
    plt.subplot(2, 5, i+1)
    probplot(normalize_rtn_1d[col[i]], dist="norm", plot=plt)
    plt.xlabel('Normal Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.title(col[i])
    plt.grid()

plt.show()
```

극단값 몇 개를 제외하고 정규성을 가지고 있는 것으로 보입니다.

### Problem. 1 Two correlated assets


| Asset | r̅    | σ   |
|-------|-------|-----|
| A     | 10.0% | 15% |
| B     | 18.0% | 30% |

The correlation between assets A and B is 0.1, and other data are given in Table 1. (Note: $\displaystyle \rho = \frac{\sigma_{AB}}{\sigma_A \sigma_B}$)

**(a) Find the proportions α of A and (1 - α) of B that define a portfolio of A and B having minimum standard deviation.**

```python
r_A = 0.10
r_B = 0.18

sigma_A = 0.15
sigma_B = 0.30

corr_AB = 0.10

w_A = (sigma_B**2 - (sigma_A * sigma_B * corr_AB)) / (sigma_A**2 + sigma_B**2 - (2 * sigma_A * sigma_B * corr_AB))
w_B = 1 - w_A

print(f'A: {w_A:.2%}')
print(f'B: {w_B:.2%}')
```

**(b) What is the value of this minimum standard deviation?**

```python
var_p = ((w_A**2) * (sigma_A**2)) + ((w_B**2) * (sigma_B**2)) + (2 * w_A * w_B * sigma_A * sigma_B * corr_AB)
std_p = np.sqrt(var_p)

print(f'Minimum Standard Deviation: {std_p:.4f}')
```

**(c) What is the expected return of this portfolio?**

```python
r_p = (r_A * w_A) + (r_B * w_B)
print(f'Expected Return: {r_p:.2f}')
```
