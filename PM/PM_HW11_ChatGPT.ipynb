## Summarize the paper

- Can ChatGPT Improve Investment Decision (고형진, 이재욱)

### ABSTRACT

LLM으로서 ChatGPT는 큰 관심을 받고 있는데, 투자 분야에서의 GPT 잠재력은 아직 밝혀지지 않았다.

목표: 정량적 방법론을 이용한 자산 배분 포트폴리오 관리를 지원하는 GPT의 효과를 조사

- GPT를 이용하여 정해진 클래스를 선택하고, 그 선택의 분산 효과를 평가

결과: GPT가 선택한 자산은 무작위로 선택한 자산보다 diversity index 측면에서 통계적으로 우수하다.

- 포트폴리오 매니저의 보조로써 GPT 로보어드바이저 역할에 대한 인식 개선에 기여

### 1. Introduction

GPT(Generative Pre-trained Transformer)에 대한 관심

- 출시 첫 주 100만명 이상 가입
- 월간 웹사이트 방문자 수: 6억 1,600만명

LLM으로서 GPT가 사람이 포착할 수 없는 시장의 특수한 측면을 포착할 수 있을까?

- 자산 선택, 포트폴리오 다각화 달성에 도움을 줄 수 있는지 조사
    - GPT는 예측 모델이 아님
- GPT에게 질문을 던져 자산을 선택하고, 3가지 diversity index를 계산하여 자산을 평가
- 이후, 해당 포트폴리오와 랜덤으로 선택한 자산으로 구축된 포트폴리오 성과를 비교

결과

- 랜덤 선택보다 통계적으로 유의미하게 개선되어 GPT의 선택이 자산 배분 철학에 기반하고 있음을 입증
- GPT 포트폴리오 성과가 랜덤 포트폴리오 성과를 능가
- GPT 선택 자산은 충분한 분산 효과를 가져와 보다 효율적인 포트폴리오를 구축할 수 있다.

### 2. Methodology

#### 2-1. Diversity indices

- Simpson and Shannon diversity indices
    - 고정된 수의 자산이 선택되는 시나리오 평가에 적합  
    
    $$I_{\text{Simpson}} = 1 - \sum_{i=1}^{s} p_i^2$$  
    
    $$I_{\text{Shannon}} = - \sum_{i=1}^{s} p_i \ln(p_i)$$  
    
    - $p_i$: 전체 자산에서 i번째 자산의 비율
    - $s$: 자산 수

- The average correlation

    $$\rho_{\text{avg}} = \frac{2}{k(k-1)} \sum_{i=1}^{k-1} \sum_{j=i+1}^{k} \rho_{ij}$$
        
    - $p_{i,j}$: 자산 i와 자산 j의 상관계수
    - $k$: 자산 개수

#### 2-2. Portfolio analysis based on the mean-variance framework

$$\begin{align*}
& \min_{w} \quad \frac{\lambda}{2} w^T \Sigma w - w^T \mu, \\
& \text{s.t.} \quad w^T \mathbf{1} = 1, \\
& w_i \geq 0 \quad (i = 0, \ldots, K)
\end{align*}$$  

MVO를 이용해 최적 포트폴리오 구성

- 이후 포트폴리오 성과를 평가하여 GPT의 의사결정을 통해 diversification 정도 평가

요약하면

- diversity index와 평균 상관계수 지표를 이용하여 다양한 자산을 선별하는 GPT 능력을 평가하고
- MVO를 이용해 달성한 diversification 정도를 평가

이를 통해 GPT의 보조자 역할 능력을 확인

### 3. Data and experimental design

#### 3-1. Data

기간: 2021년 1월 1일 ~ 2023년 1월 31일  

총 40개의 자산 수집

- 미국 주식 시장에 상장된 다양한 섹터의 대형주 20개
- 암호화폐 5개
- 원자재 5개
- 통화 5개
- 채권 5개  

일별 단순 수익률 계산 (IS: 252개, OS: 274개)

T-bill로 무위험 수익률 대체

#### 3-2. Experimental design

- 5개 자산 클래스의 40개 자산 중 k개의 자산을 선택하도록 GPT에 요청
- 무작위로 k개 자산 선택을 반복
- diveristy index를 계산하고, 포트폴리오 최적화를 진행
- OS에 대해 포트폴리오의 SR과 리스크를 평가
- 해당 과정 10,000번 반복
- IS: 2021년 1월 1일 ~ 2021년 12월 31일
    - 예상 수익률 및 공분산 추정
    - k는 짧은 시계열을 고려해 2, 3, 4로 설정
    - GPT는 21년 12월 31일까지 데이터로 학습된 모델
- OS: 2022년 1월 1일 ~ 2023년 1월 31일

### 4. Empirical Findings

- 자산 개수에 관계없이 모두 성과가 랜덤 셀렉션보다 높고, t-value도 유의하다.

- MVO로 최적화
- 수익률과 SR 역시 통계적으로 유의하게 좋은 성과를 보이고 있다.
- 또한, GPT 방식은 자산이 늘어감에 따라 SR도 좋아지고 있다.

- GMV로 최적화
- 랜덤 셀렉션보다 낮은 리스크를 얻을 수 있다.
- k=2일 때, GPT의 SR이 상대적으로 떨어지는데, 이는 MVO가 아니기 때문

결론적으로 GPT의 선택은 포트폴리오 구성에서 분산 효과를 제공하고, 효율적인 리스크 관리를 달성

### 5. Conclusion

- GPT는 투자 과정에서 유용한 보조 도구가 될 수 있다.
- GPT가 매니저와 같은 수준으로 미래를 예측하거나 포트폴리오를 최적화할 순 없지만, 해당 지식과 숙련도가 부족한 개인에게 편리하고 실용적인 옵션을 제공한다.
- 이러한 개인들에게 교육자료로 활용될 수 있다.
- GPT는 금융 산업의 예언가가 아니라 조력자이다.
