```python
# 위 프로세스 함수로 정리

from scipy.optimize import minimize

def find_a(mkt_spot, b):
    def obj_function(a):
        short_lattice = pd.DataFrame(columns = [ i for i in range(14)], index = [ i for i in range(13, -1, -1) ])
        short_lattice.loc[0,0] = a[0]
        for t in range(1, len(short_lattice.columns)):
            for s in range(t+1):
                short_lattice.loc[s,t] = a[t]*np.exp(b*s)

        el_price = pd.DataFrame(columns = [ i for i in range(15)], index = [ i for i in range(14, -1, -1) ])
        el_price.loc[0,0] = 1
        for t in range(1, len(el_price.columns)):
            for s in range(t+1):
                if s == t:
                    el_price.loc[s,t] = 0.5 * el_price.loc[s-1,t-1] / (1 + short_lattice.loc[s-1,t-1]/100)
                elif s == 0:
                    el_price.loc[s,t] = 0.5 * el_price.loc[s,t-1] / (1 + short_lattice.loc[s,t-1]/100)
                else:
                    el_price.loc[s,t] = 0.5*((el_price.loc[s-1,t-1]/(1+short_lattice.loc[s-1,t-1]/100))+(el_price.loc[s,t-1]/(1+short_lattice.loc[s,t-1]/100)))

        bdt_zcb = el_price.sum(axis=0)
        bdt_spot = 100 * (((1/bdt_zcb)**(1/bdt_zcb.index))-1)

        # 차의 제곱의 합이 최소가 되도록 하는 a를 구하자.
        return ((bdt_spot.iloc[1:] - mkt_spot)**2).sum()

    a_init = [ 5 for i in range(14)]
    b_ = [(0,20) for i in range(len(a_init))]

    result = minimize(obj_function, a_init, bounds = b_)
    return result.x

def swaption_pricing(a, b, e):

    short_lattice = pd.DataFrame(columns = [ i for i in range(14)], index = [ i for i in range(13, -1, -1) ])
    short_lattice.loc[0,0] = a[0]
    for t in range(1, len(short_lattice.columns)):
        for s in range(t+1):
            short_lattice.loc[s,t] = a[t]*np.exp(b*s)

    short_lattice = short_lattice/100
    fix = 0.1165

    swap_price = short_lattice.loc[9:,:9].copy()
    swap_price.iloc[:,-1] = (swap_price.iloc[:,-1]-fix) / (1+swap_price.iloc[:,-1])
    for t in swap_price.columns[:-1][::-1]:
        for s in range(t+1):
            swap_price.loc[s,t] = 1/(1+short_lattice.loc[s,t])*((short_lattice.loc[s,t]-fix) + 0.5*(swap_price.loc[s+1,t+1] + swap_price.loc[s, t+1]))

    swaption_price = swap_price.copy()
    swaption_price[swaption_price<0] = 0
    swaption_price = swaption_price.loc[e:,:e]
    for t in swaption_price.columns[:-1][::-1]:
        for s in range(t+1):
            swaption_price.loc[s,t] = (swaption_price.loc[s+1,t+1] + swaption_price.loc[s,t+1])*0.5/(1+short_lattice.loc[s,t])
    
    return swaption_price.loc[0,0]*1000000
```

## Problem 3.

Construct a short lattice for periods (years) 0 through 9 with an initial rate of 6% and with successive rates determined by a multiplicative factor of either $u = 1.2$ or $d = 0.9$. Assign the risk-neutral probabilities to be 0.5.    

```python
import pandas as pd
import numpy as np

u = 1.2
d = 0.9
q = 0.5
initial_rate = 0.06 
period = 10
```

```python
df = pd.DataFrame(np.diag([initial_rate * d**i for i in range(period)]))
for i in range(period):
    for j in range(period):
        if j > i:
            df.iloc[i,j] = df.iloc[i,j-1]*1.2
df
```

**(a) Using this lattice, find the value of a 10-year 6% bond.**

```python
face_value = 100
c = face_value*0.06

df_pv = pd.DataFrame().reindex_like(df)
df_pv.iloc[:,-1] = (face_value + c) / (1+df.iloc[:,-1])

for i in range(period-2, -1, -1):
    for j in range(period):
        if i >= j:
            p_u = df_pv.iloc[j,i+1]
            p_d = df_pv.iloc[j+1,i+1]
            pv = (q*p_u + (1-q)*p_d)
            df_pv.iloc[j,i] = (pv + c) / (1+df.iloc[j,i])
df_pv
```

쿠폰 금리 6%, 만기 10년 채권의 현재가치는 약 **91.7194$**이다.

**(b) Suppose this bond can be called by the issuing party at any time after 5 years. (When the bond is called, the face value plus the currently due coupon are paid at that time and the bond is canceled.) What is the fair value of this bond?**

```python
df_pv_call = pd.DataFrame().reindex_like(df)
df_pv_call.iloc[:,-1] = (face_value + c) / (1+df.iloc[:,-1])
df_call = df.copy()

for i in range(period-2, -1, -1):
    for j in range(period):
        if i >= j:
            if (df_call.iloc[j,i] < 0.06) & (i > 4):
                df_call.iloc[j,i] = 0.06
            p_u = df_pv_call.iloc[j,i+1]
            p_d = df_pv_call.iloc[j+1,i+1]
            pv = (q*p_u + (1-q)*p_d)
            df_pv_call.iloc[j,i] = (pv + c) / (1+df_call.iloc[j,i])
df_pv_call
```

단기 이자율이 쿠폰 이자율인 6%보다 낮으면 발행자는 채권을 회수하는 것이 유리하다.  

따라서 5년 후 단기 이자율이 6% 이하인 경우를 감안하여 계산하면 현재가치는 약 **90.9560$**이다. 

**(c) Use the forward equation to find the spot rate curve for the lattice.**  

```python
df_forward = pd.DataFrame().reindex_like(df)
df_forward.iloc[0,0] = 1

for s in range(period):
    for k in range(s, period):
        if k == 0:
            continue
        elif s == 0:
            df_forward.iloc[s,k] = (q*df_forward.iloc[s, k-1]) / (1+df.iloc[s,k-1])
        elif s == k:
            df_forward.iloc[s,k] = (q*df_forward.iloc[s-1, k-1]) / (1+df.iloc[s-1,k-1])
        else:
            df_forward.iloc[s,k] = q*((df_forward.iloc[s-1, k-1])/ (1+df.iloc[s-1,k-1]) + df_forward.iloc[s,k-1]/(1+df.iloc[s,k-1]))
df_forward
```

```python
bond_price = df_forward.sum(axis=0)
bond_price
```

```python
spot_rates = (1/bond_price)**(1/bond_price.index) - 1
spot_rates
```

## Problem 4.

What is the expected number of cards that need to be turned over in a regular 52-card deck in order to see the first ace?


에이스를 제외한 카드를 1~48이라 할 때,

$X_i =
\begin{cases}
1,\;\text {if card i is turned over before 4 aces}\\
0,\;\text {otherwise}
\end{cases}$

첫번째 A가 나올 때까지 뽑아야하는 카드 수 $X = 1 + \displaystyle \sum_{i=1}^{48}X$이고, 기댓값 $E[X] = 1 + \displaystyle \sum_{i=1}^{48}E[X_i]$ 이다.  

A 4장을 기준으로 카드 전체를 5구간으로 나누면 A 이전 구간에서 카드를 뽑을 확률은  $\displaystyle \frac {1}{5}$ 이다.

따라서 $E[X] = 1 + \displaystyle \sum_{i=1}^{48}E[X_i] = 1 + \frac {48} {5} = 10.6$

## Problem 5.

Solve the corresponding leetcode problem below and register the solution on GitHub.

**two sum**  

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.  
You may assume that each input would have exactly one solution, and you may not use the same element twice.  
You can return the answer in any order.

```python
nums = [2,7,11,15]
target = 9
```

```python
class Solution5:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i] + nums[j] == target:
                    return [i,j]
```

## Problem 6.

Solve the corresponding leetcode problem below and register the solution on GitHub.

**Valid Palindrome**

A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise.

```python
s = "A man, a plan, a canal: Panama"
```

```python
class Solution6:
    def isPalindrome(self, s: str) -> bool:
        result = ''
        for c in s:
            if not (c.isalpha() or c.isnumeric()):
                continue
            elif c.isupper():
                result += c.lower()
            else:
                result += c
        return result == result[::-1]
```

```python
Solution6().isPalindrome(s)
```
