### 10.3 The APT, the CAPM, and the Index Model

- APT는 잘 분산되지 않은 포트폴리오에도 적용되는가?
- 리스크와 수익률 모형으로서 APT는 CAPM과 비교하여 우월한가?
- 증권분석 애널리스트가 약간의 잔차 리스크가 남아있는 양의 알파를 가진 포트폴리오를 찾아냈다고 하자. 우리는 이미 지수 모형에 적용한 Treynor-Black 방법으로부터 이 경우에 대한 해결책을 가지고 있지 않은가? 이런 분석의 틀이 APT보다 나은가?

CAPM은 거의 모든 투자자가 MVO를 추구한다는 조건인데 APT는 소수의 차익거래자들이 차익거래 기회가 있는 시장을 찾아다니기만 하면 충족된다. (이거만으로 수익률 = 베타관계식 형성) 즉, 잔차만 괜찮으면 된다.  

CAPM은 관찰할 수 없는 모든 자산을 포함하는 포트폴리오에 의존하고 검증조차 불가능하지만, APT는 시장지수와 같은 관찰이 가능한 포트폴리오에 기반을 두고 있다.  

단, APT가 CAPM보다 우월하지 못한 이유는 개별 자산 수준이나 잔차 리스크 수준에서 순수한 차익거래만으로는 수익률-베타 관계식을 강제하기에 불충분하기 때문이다.  

**결론: 두 모형 어느 것도 한계로부터 자유로울 수 없다.** 

APT 문제점:
차익거래 포트폴리오가 진짜로 무리스크여야한다. 차익거래 포지션이 완전히 잘 분산되지 않으면 그 규모를 확대하는 것 차익거래 포지션의 리스크를 잠재적으로는 무한정 증가시킬 것이다. APT는 이러한 문제를 무시한다.  

최적화 과정: 가격오류가 있는 자산에 대한 포지션에서 얻는 잠재적 이익뿐 아니라 전체 포트폴리오의 리스크와 효율적 분산투자 모두 고려.  
Treynor-Black(T-B)
1. 벤치 마크 포트폴리오의 리스크 프리미엄과 표준편차를 추정한다.  
2. 가격오류가 있는 모든 자산을 액티브 포트폴리오에 편입한다. 이 포트폴리오의 알파를 $\alpha_A$, 체계적 리스크 계수를 $\beta_A$, 잔차리스크를 $\sigma(e_A)$라 할 때, 액티브 포트의 투자 비율은
$$ w_A^0 = \frac {\alpha_A/\sigma^2(e_A)} {E(R_M) / \sigma^2_M},　 w^*_A = \frac {w^0_A} {1 + W^0_A(1-\beta_A)}$$  
최적화 포트폴리오의 샤프비율은 $\sqrt {S^2_M + IR^2_A }$,  $S_M$은 패시브 포트 샤프, $IR$은 $\sigma_A/\sigma(e_A)$
3. 리스크 포트폴리오의 샤프 비율을 극대화하려면 적극적 포트폴리오의 IR을 극대화하면 된다. 이는 알파가 0이 아닌 각 자산 $w_Ai = \sigma_i/\sigma(e_A)$에 비례하여 포트폴리오 투자 비중을 배분함으로써 달성한다.

### 10.5 The Fama-French (FF) Three-Factor Model

# [Problem Solving]

**The analyst wants to evaluate the performance of Berkshire Hathaway from Jan/1999 to Dec/2013. Determine whether the fund has generated alpha (i.e., excess returns that cannot be explained by exposure to some factors) using returns of the market, the size factor (SMB), the value factor (HML), the quality factor (QMJ), the low volatility factor (BAB), and the momentum factor (UMD).**

```python
# 데이터 준비
import pandas as pd
import statsmodels.api as sm

data_path = 'data'
df = pd.read_csv(data_path + '\BRK13F.csv')
df = sm.add_constant(df)
df.rename(columns = {'const':'ALPHA'}, inplace = True)
f_col = ['BRK.A','MKT','SMB','HML','QMJ','BAB','UMD']
df[f_col] = df[f_col].apply(lambda x: x - df['RF'])
df
```

```python
df[df['datadate'] == '1/31/2011']
```

```python
df.iloc[0,1]
```

----

**1) Analyze the performance of BRK and determine whether the fund has generated alpha with the Fama-French 3 Factor Model (MKT, SMB, HML)**

```python
# 팩터 설정
col1 = ['datadate','BRK.A','ALPHA','MKT','SMB','HML']
df1 = df[col1]

# Regression
y = df1['BRK.A']
X = df1[['ALPHA','MKT','SMB','HML']]
ff3_model = sm.OLS(y, X)
ff3_fit = ff3_model.fit()
ff3_fit.summary()
```

```python
ff3_fit.resid.plot(figsize = (10,5), title = 'residuals');
```

알파의 p-value는 0.083으로 유의수준 5%에서 알파는 0이라는 귀무가설을 채택한다.  
따라서 Fama-French 3 Factor는 버크셔의 초과수익을 충분히 설명하고 있다.  

이 모델의 수정 $R^2$는 0.615로, 약 61.5%의 설명력을 보이고 있다.

---

**2) Analyze the performance of BRK and determine whether the fund has generated alpha with the Carhart 4 Factor Model (MKT, SMB, HML, UMD)**  

```python
# 팩터 설정
col2 = ['datadate','BRK.A','ALPHA','MKT','SMB','HML','UMD']
df2 = df[col2]

# Regression
y = df2['BRK.A']
X = df2[['ALPHA','MKT','SMB','HML','UMD']]
c4_model = sm.OLS(y, X)
c4_fit = c4_model.fit()
c4_fit.summary()
```

```python
c4_fit.resid.plot(figsize = (10,5), title = 'residuals');
```

알파의 p-value는 0.048로 유의수준 5%에서 알파는 0이라는 귀무가설을 기각한다.  
따라서 Carhart 4 Factor는 버크셔의 초과수익을 설명하기 위한 개선의 여지가 있다.  

이 모델의 수정 $R^2$는 0.634로, 약 63.4%의 설명력을 보이고 있다.
