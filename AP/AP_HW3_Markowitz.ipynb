### 7.2 두개의 리스크 자산으로 구성된 포트폴리오

두 자산의 공분산 항이 음인 경우 포트폴리오 분산이 감소

공분산이 양의 값을 가지는 경우에도 두 증권이 완전한 양의 상관관계를 가지지 않는 한 포트폴리오 표준편차는 각 증권의 표준편차의 가중평균치보다 작음


위 식에 따라 상관계수가 음일 때 분산은 감소. 양이면 분산이 증가, 1이면 개별 증권의 가중평균과 같음

따라서 증권 간의 상관관계가 작을수록 포트폴리오의 효율성은 향상 (분산투자는 효과적이고 리스크는 낮아짐)

기대수익률은 상관관계에 영향을 받지 않고, 분산은 받음


상관관계가 -1이라면 위 식으로 완전 헷지 포지션을 얻을 수 있음

한쪽 자산에 쏠릴수록 분산투자 효과가 사라져 다시 표준편차가 올라감

최소분산 포트폴리오의 표준편차는 개별 증권의 표준편차보다 낮음

비체계적 위험을 분산시킴으로써 단순히 가중평균만큼의 이득이 아닌 더 큰 이득을 얻음 (상관계수에 따라)

기대수익률은 가중평균과 같지만 포트폴리오의 표준편차는 그렇지 않기 때문

우리가 평소에 "리스크가 큰만큼 수익률도 크다!"라고 하지만, '하이 리스크. 하이 리턴'은 사실 정확한 비례관계가 아닌 것
(분산투자를 통해 수익률을 올리면서 리스크는 그보다 덜 올릴 수 있음)

```python
import pandas as pd
import FinanceDataReader as fdr
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['axes.unicode_minus'] = False
```

### 2 - (a)

```python
# T-bill
tbill = fdr.DataReader('TB3MS', data_source = 'fred', start = '2013-01-01', end = '2022-12-31')
tbill.index = tbill.index.strftime('%Y-%m')

# Stock
stocks_d = yf.download(tickers = 'SPY AAPL', start = '2013-01-01', end = '2023-01-01').loc['2013-01-01':,'Adj Close']
stocks_d.index = stocks_d.index.strftime('%Y-%m')
stocks_m = stocks_d.groupby('Date').apply(lambda df: (df.iloc[-1,:] / df.iloc[0,:]) - 1)

# Merge
df_a = pd.concat([tbill, stocks_m], axis=1)

# Plot
df_a.plot(kind = 'scatter', x = 'SPY', y = 'AAPL', title = 'SPY - AAPL', figsize = (10,6));
```

### 2 - (b)

```python
df_a['mk_rtn'] = df_a['SPY'] - df_a['TB3MS']
X = sm.add_constant(df_a['mk_rtn']) 
y = df_a['AAPL'] - df_a['TB3MS']

model = sm.OLS(y, X)
results = model.fit()

print(f'alpha : {results.params[0]:.4f}')
print(f'beta : {results.params[1]:.4f}')

results.summary()
```

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
```

### 3 - (a)

```python
# 데이터 다운로드

rtn = yf.download('AAPL MSFT AMZN GOOGL NVDA', start = '2018-01-01', end = '2023-01-01').loc[:,'Adj Close']
rtn_1d = rtn.pct_change(1, axis=0).iloc[1:]
rtn_1d
```

### 3 - (b)

```python
# 연평균 수익률 및 공분산 행렬

annual_rtn = rtn_1d.mean(axis=0) * 252
annual_cov = rtn_1d.cov() * 252
```

```python
annual_rtn
```

```python
annual_cov
```

### 3 - (c)

```python
# 몬테카를로 시뮬레이션

port_ret = []
port_risk = []
port_weights = []

for _ in range(20000):
    weights = np.random.random(rtn_1d.shape[1])
    weights /= np.sum(weights)
    
    temp_returns = np.dot(weights,annual_rtn)
    temp_risk = np.sqrt(np.dot(weights.T, np.dot(annual_cov, weights)))
    
    port_ret.append(temp_returns)
    port_risk.append(temp_risk)
    port_weights.append(weights)
    
pf = pd.DataFrame({'Return': port_ret, 'Risk': port_risk, 'Weight': port_weights})
pf['Sharpe'] = ((pf['Return'] - 0.02) / pf['Risk'])
pf.plot(kind = 'scatter', x = 'Risk', y = 'Return', title = 'Risk - Return',
        figsize = (10,6), c = 'Sharpe', cmap = 'viridis', edgecolor='k');
```

### 3 - (d)

```python
# 가장 높은 샤프로 CAL Plot

max_sharpe = pf['Sharpe'].max()
max_sharpe_pf = pf.loc[pf['Sharpe'].idxmax(),:]

x = np.array([-0.5, 0.5])
y = 0.02 + x * max_sharpe
cal = pd.Series(index = x, data=y)

pf.plot(kind = 'scatter', x = 'Risk', y = 'Return', figsize = (10,6), c = 'Sharpe', cmap = 'viridis', edgecolor='k', grid = True);
cal.plot(kind = 'line', xlim = (0.27, 0.45), ylim = (0.10, 0.35), label = 'CAL');
plt.scatter(max_sharpe_pf['Risk'], max_sharpe_pf['Return'], c='b',
           marker='o', s=100, label = 'Max Sharpe Portfolio')
plt.title('Optimal Portfolio - CAL');
```
