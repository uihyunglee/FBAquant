## Chapter 9. Data and Statistics

### 9.1 Basic Estimation Methods

MVO에서 중요한 것은 파라미터를 추정하는 것: 기대수익률, 분산, 공분산  

전통적 방법: 과거 데이터 이용  
단, 분산이나 공분산은 어느정도 믿을 수 있지만 기대수익률 같은 다른 파라미터는 믿을 수 없다.  
이는 자료나 계산이 잘못된 것이 아닌 그저 추정값의 근본적인 한계 때문 (blur of history)  

**Period-Length Effects**  

$\bar{r}_p = p\bar{r}_y$  
$\bar{\sigma}_p = \sqrt{p}\bar{\sigma}_y$  

단기간 파라미터를 장기간 파라미터로 근사할 수 있다.  

수익률은 기간에 직접적으로 비례하고, 표준편차는 기간 제곱근에 비례하기 때문에 기간이 감소함에 따라 표준편차/기대수익률 비는 증가한다.  
따라서 단기간 수익률은 기댓값에 비해 표준편차가 매우 높다. (기간을 길게 잡으면 샤프가 끝도 없이 커지는게 이거 때문)  

주식의 연평균 수익률은 6\~30%이고 12%가 일반적 (이는 시간에 따라 변하기 때문에 2-3년 동안만 의미가 있음)  
주식의 연평균 표준편차는 10~60%이고 15%가 일반적  
이에 따라 계산해보면  
연: 표준편차는 수익의 1.25 배  
월: 표준편차는 수익에 4.3 배  
일: 표준편차는 수익의 19.8 배  
(성과 분석 시 주의)  

**Mean Blur**  

증폭효과로 인해 기대 비율 추측값을 만들기가 불가능  

과거 자료를 가지고 정확한 $\bar{r}$을 찾기는 기본적으로 불가능

### 9.3 The Effect of Estimation Errors

포트폴리오 구축은 파라미터 추정치에 의존하는데, 그건 정확하지 않다.  

악영향은 특히 기대수익률에 집중  

**추정치 오류에 대한 부정적 영향** 
- Condition number: poorly conditioned covariance matrix는 $\bar{r}$의 추정치 오류를 증가시키고 포트폴리오 가중치를 잘못 추출한다.
- Leverage: 레버리지는 추정치 오류에 더 민감하다. 
- Large weights: 높은 가중치의 자산이 있다면, 전반적인 수익률은 $\bar{r}$에 의해 더 강하게 결정된다.  

**Three Views**

추정 오류는 최적의 포트폴리오를 만들지 못하게 한다. 

think: 부정확한 추정치를 사용한 커브
theory: 실제 파라미터를 사용한 커브
actual: 기대수익률 추정치를 이용해 계산한 커브

### 9.4 Conservative Approaches

레버리지, 패널티를 통한 가중치 억제, MSE 등으로 보수적 접근을 해볼 필요가 있다.  

결국 가장 중요한 것은 정확한 기대수익률을 찾는 것

### 9.5 Tilting Away From Equilibrium

CAPM은 가장 효율적인 포트폴리오가 시장 포트폴리오가 되며 복잡한 계산을 하지 않고 단순히 시장 포트폴리오와 무위험 자산을 통한 효율적 투자를 가능하게 한다.  

이와 반대로 대부분의 투자자는 적절한 파라미터를 적용한 마코위츠 모델로 시장 포트폴리오보다 더 나은 포트폴리오가 있다고 믿는다.  
하지만 과거데이터를 이용한 파라미터 추정은 의미가 없기 때문에 더 효과적인 추정 방법이 필요하다.  

과거 전망이 아닌 미래 전망을 통한 정보
- 회사에 대한 기본적 분석, 미래 프로젝트에 대한 분석, 재정상태, 경쟁기업, 상품과 시장에 대한 분석
- 여러 분석가의 결론
- 기사 또는 개인적 경험에 근거한 예감이나 직관
