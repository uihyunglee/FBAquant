## Chapter 5: Risk, Return, and the Historical Record  
### 5-1. 이자율 수준의 결정 요인  
금리는 거시경제의 중요한 팩터이며 직접적으로 수익률과 연관이 있다. 보통 금리가 오르면 주식시장에는 악영향을 미치고, 내리면 긍정적인 영향을 미친다. 따라서 금리를 예상하는 것은 중요하지만 이는 매우 어렵다. 다만, 아래의 요인들이 금리 수준을 결정한다고 알려져있다.  
- 저축으로부터의 자금 공급양
- 기업으로부터의 자금 수요 (시설, 장비 등 실물자산과 자본 축적을 위해)
- 정부(연준)의 수요 
- 기대 인플레이션  

근본적인 실질금리의 결정요인은 세대 저축 성향, 기업의 물리적 자본에 대한 투자의 기대 이익성이라고 볼 수 있지만, 정부의 재정, 화폐정책의 영향을 받을 수 있다.

금리가 올라갈수록 저축률은 높아진다. 금리가 오르면 세대는 현재 소비를 꺼릴 것이고, 가용 자산을 유보하거나 미래를 위해 투자할 것이다. 반대로 금리가 떨어지면 많은 기업이 자본을 높이기 위해 더 많이 투자하고 싶어할 것이다. 실질금리의 균형은 수요와 공급, 두 곡선이 만나는 지점에 생긴다. 여기서 정부나 중앙은행의 재정, 화폐 정책으로 이 곡선을 옮길 있다.  
  
**정부의 예산 적자 -> 정부 차용 수요 증가 -> 수요 곡선 오른쪽으로 이동 -> 균형 실질 이자율 $E'$로 증가**

**이때 연준은 확장 통화 정책을 통해 이를 상쇄 -> 공급 곡선 오른쪽으로 이동 ->균형 실질 이자율 유지**
    
그리고, 금리에 영향을 미치는 요인으로 인플레이션 또한 중요하다. 우리가 무위험 금리라 부르는 이자 또한 인플레이션이 높다면 무위험이 아닐 수 있다. 이는 인플레이션이 명목금리를 넘어선다면 실질금리는 음수로 가기 때문이다.  

**명목금리가 10%라 해도, 인플레이션이 6%라면 결국 4%밖에 남지 않는다.**  

정확한 공식과 근사식은 아래와 같다. 정확한 공식은 후에 연속복리로 다시 정의된다.

### 5-2. 보유기간이 다른 수익률의 비교
**연간 실효이자율 (effective annual rate: EAR)**: 1년 간 투자된 자금의 증가율이며, 1년보다 단기이면 복리로 계산.  
**연간 이자율 (annual percentage rates: APR)**: 1년 미만 단기 투자 수익률은 복리가 아니라 단리로 연율화하기도 함. 제곱이 아닌 일반곱으로 계산.  
**연속복리**: 기간 T가 0에 근접하면 실질적으로 연속복리에 접근. 이를 $r_{cc}$라고 함. 이를 이용하면 실질금리 식은 정확한 식이 됨. 복리계산 빈도가 적을수록 동등한 실효이자율을 얻으려면 더 높은 APR이 필요.

### 5-3. 재정증권과 인플레이션
높은 인플레이션으로 인해 1970년대부터 평균 이자율(명목)은 그 전보다 상승 -> 1970년 초부터 2012년까지 T-bill에 투자했다면 약 10배 상승  

그러나, 실질금리로 따보면 결국 20% 상승이었다. 결국 실질 수익률이 가장 중요하다. 높은 인플레이션이 예상될 때는 더 높은 명목 수익률이 필요하며 어떤 투자기간 초기의 T-bill 명목수익률은 그 기간 동안의 예상 인플레이션을 반영해야한다.  

**이상적인 상황에서 인플레이션과 명목 이자율의 상관계수는 1에 가까울 것이며, 인플레이션과 실질 이자율과의 상관계수는 0에 가까울 것이다.  
투자자들이 인플레이션을 무시하거나 예상 능력이 전혀 없다면 명목 이자율과 T-bill의 상관계수는 0, 인플레이션과는 -1일 것이다.**

인플레이션 예상 정확도가 상승하며, 최근으로 올수록 전자의 환경에 가까워지고 있다.

### 5-4. 리스크와 리스크 프리미엄

**보유기간 수익률 (holding period return: HPR)**: 시세차익 + 배당 수익  
**기대수익률 예측**: 시나리오에 대한 확률과 HPR을 가중평균 계산하여 예측. 리스크는 예상 수익률과 시나리오 수익률의 표준편차로 계산.  
확률분포가 정규분포이면 기대수익률과 표준편차로 확률분포 상황을 완벽히 나타낼 수 있다.  

**리스크 프리미엄 = 수익 - 무위험 수익**

### 5-5. 과거 수익률의 시계열 분석

**시나리오 분석**: 시나리오 구상 - 각 투자수익률 결정 - 각 확률 추정 - 프리미엄, 리스크 계산. 역사적 실적 자료 이용 시 발생확률은 같다고 가정(산술평균)  

**시계열 분석**: 시기별 HPR 관측 - 수익률 발생 추정 구간의 확률분포, 기대수익률, 표준편차 추정  


**기하평균(시간가중평균) 수익률**: 전체 누적수익률 중 연 수익률(HPR) (복리계산). 단순산술평균은 투자금이 변하며 생기는 손익을 잡아내지 못함.

수익률의 변동폭이 클수록 기하평균과 산술평균의 차이는 커진다. (복리 수익률과 연간 수익률의 평균치)

**리스크**: 기대수익률의 추정치인 산술평균 $\bar r$ 로부터 편차의 제곱을 평균하여 분산을 추정

과거 수익률 성과에 대해서는 기하평균, 미래 수익률 예상을 예상할 때는 산술 평균을 이용

**고빈도 관측치의 평균과 표준편차 추정**
- 수익률 추정의 정확도는 관측의 수가 아닌 표본 기간의 길이에 영향을 받는다.
- 평균치와 달리 표준편차나 그 이상의 모멘트는 관측 수를 늘림으로써 정확도를 높일 수 있다.

**보상-변동성 비율Sharpe Ratio**

## Chapter 6: Capital Allocation to Risky Assets

### 6.1 리스크와 리스크 회피

도박: 리스크를 짊어지는 행위 자체를 즐기는 것  
투기: 리스크에 상응하는 이익을 얻기 위한 것  

리스크회피적 투자자는 무리스크, 리스크 프리미엄이 양인 투기적 투자만 고려  

U(효용가치) = 기대수익률 - 1/2 * 위험 회피도(A) * 분산  

확정등가 수익률(centainty equivalent rate): 리스크 포트폴리오와 동일한 효용 점수를 가지기 위해 무리스크 투자가 제공해야하는 수익률  
-> 무리스크 확정등가 수익률보다 효용이 높을 때만 그 포트폴리오는 바람직하다.  

충분히 리스크 회피적인 투자자는 양의 리스크 프리미엄을 가진 경우에도 무리스크보다 낮은 확정등가 수익률을 부여하며, 리스크 포트폴리오를 거부한다.  

리스크 중립적 투자자(A=0)는 기대수익률에 의해서만 투자  
리스크 선호자는 공정게임이나 도박을 즐긴다.  

무차별곡선(indifference curve): 동일한 효용가치를 가지는 모든 포트폴리오의 점은 연결한 선  
-> A가 높다면 위험 대비 수익률이 높아야 효용이 같아지므로 가파르게 상승


### 6.2 리스크 포트폴리오와 무리스크 포트폴리오 간의 자산배분

### 3 - (a)

```python
# import
import pandas as pd

#데이터 읽기
df_3a = pd.read_csv('6_Portfolios_2x3.csv', header = 9)
df_3a.rename(columns={df_3a.iloc[:, 0].name:'date'}, inplace=True)

# 가치가중 수익률 선택
df_3a = df_3a.loc[:df_3a[df_3a['date'].str.isnumeric() == False].index[0]-1, :].query('(date >= "193001")&(date <= "201612")')

# 전처리
df_3a.index = df_3a.pop('date')
df_3a = df_3a.astype(float)

# 데이터 반으로 나누기
df_3a_A = df_3a.iloc[:int(len(df_3a)/2), :]
df_3a_B = df_3a.iloc[int(len(df_3a)/2):, :]

# 평균, 표준편차, 왜도, 첨도 계산
dict_3a_A = dict()
dict_3a_B = dict()

dict_3a_A['A_mean'] = df_3a_A.mean()
dict_3a_A['A_std'] = df_3a_A.std()
dict_3a_A['A_skew'] = df_3a_A.skew()
dict_3a_A['A_kurt'] = df_3a_A.kurt()
dict_3a_B['B_mean'] = df_3a_B.mean()
dict_3a_B['B_std'] = df_3a_B.std()
dict_3a_B['B_skew'] = df_3a_B.skew()
dict_3a_B['B_kurt'] = df_3a_B.kurt()

result_3a_A = pd.DataFrame(dict_3a_A).transpose()
result_3a_B = pd.DataFrame(dict_3a_B).transpose()
```

### 3 - (b)

```python
# import
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 그래프 그려보기
plt.figure(figsize=(10,6))
plt.title('Distribution')
plt.xlabel('return')
plt.ylabel('pdf')
plt.grid()
for c in result_3a_A.columns:
    x = df_3a_A[c].sort_values()
    plt.plot(x, norm.pdf(x, loc = result_3a_A.loc['A_mean', c], scale = result_3a_A.loc['A_std', c]), label = f'A_{c}')
for c in result_3a_B.columns:
    x = df_3a_B[c].sort_values()
    plt.plot(x, norm.pdf(x, loc = result_3a_B.loc['B_mean', c], scale = result_3a_B.loc['B_std', c]), label = f'B_{c}')
plt.legend()
plt.show()
```

각각의 포트폴리오는 어느정도 정규분포를 따르지만 왜도와 첨도에서 차이가 있는 것으로 보여진다.

### 4 - (a)

```python
fund_rtn = 0.18
fund_risk = 0.28
tbill_rtn = 0.08

pf_rtn = (0.7 * fund_rtn) + (0.3 * tbill_rtn)
pf_risk = (0.7 * fund_risk)
print(f'Return: {pf_rtn:.2%}')
print(f'Risk: {pf_risk:.2%}')
```

### 4 - (b)

```python
my_sharpe = (fund_rtn - tbill_rtn) / fund_risk
client_sharpe = (pf_rtn - tbill_rtn) / pf_risk
print(f'My Sharpe Ratio: {my_sharpe:.2f}')
print(f'Client Sharpe Ratio: {client_sharpe:.2}')
```

### 4 - (c)

```python
slope = my_sharpe
x = np.random.randint(10, size=100)
y = tbill_rtn + slope * x
print(f'CAL Slope: {slope}')
cal = pd.Series(index = x, data=y)
cal.plot(figsize = (10,5), title = 'CAL', xlabel = 'STD', ylabel = 'Return', grid = True, xlim = (0,0.8), ylim = (0,0.8));
```

### 4 - (d)

```python
expected_rtn = 0.16
proportion = (expected_rtn - tbill_rtn) / (fund_rtn - tbill_rtn)
print(f'Risky Proportion: {proportion:.2%}')
print(f'T-bill proportion: {1 - proportion:.2%}')
```
