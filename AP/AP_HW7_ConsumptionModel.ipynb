상대가격측정
- 다른 자산 가격이 주어질 때, 해당 자산의 가치를 측정
- 블랙숄즈모형
딱 나뉘어져 있지는 않고 적당히 섞어서 이용 ex) CAPM  

Asset Pricing을 두 가지로 요약하면

$p_t = E(m_{t+1}x_{t+1})$  
$m_{t+1} = f(data, parameters)$

$p_t =$ asset price  
$x_{t+1} =$ asset payoff  
$m_{t+1} =$ stochastic discount factor

자산의 가격은 예상되는 자산 보상의 할인 가치와 같아야한다.  
실질 금리 높음 -> 저축 증가 (미래 소비 증가 기대)  
자산 가격에 대한 위험 조정: marginal utility(consumption)와 asset payoff의 공분산 (분산 X)  
consumption이 아닌 marginal utility가 근본적인 척도 (둘은 역의 관계)  

### 1.1 Basic Pricing Equation

An investor’s first-order conditions give the basic consumption-based model

### $$p_t = E_t[\beta \frac {u'(c_{t+1})} {u'(c_t)} x_{t+1}]$$

$x_{t+1} = p_{t+1} + d_{t+1}$  
$x_{t+1}$은 수익률이 아닌 $t+1$ 시점의 투자 가치  
$c_t$는 $t$시점의 소비  


#### $$U(c_t, c_{t+1}) = u(c_t) + \beta E_t[u(c_{t+1})]$$
#### $$u(c_t) = \frac {1} {1 - \gamma} c_t^{1-\gamma}$$
#### $$\gamma \to 1 : u(c) = ln(c)$$

미래를 $\beta$로 할인(impatience): subjective discount factor

#### $$max　u(c_t) + \beta E_t[u(c_{t+1})]　　s.t.$$  
$$c_t = e_t - p_t\xi$$
$$c_{t+1} = e_{t+1} + x_{t+1}\xi$$  

$e$: 기존 소비 수준(투자자가 자산을 전혀 구매하지 않은 경우), $\xi$: 구매하기로 선택한 자산의 양  

제약조건을 식에 대입하고 $\xi$에 대해 미분하면 최적의 값을 얻을 수 있다.

### $$p_t u'(c_t) = E_t[\beta u'(c_{t+1} x_{t+1})]$$
### $$p_t = E_t[\beta \frac {u'(c_{t+1})} {u'(c_t)} x_{t+1}]$$

$x_{t+1}, c_t, c_{t+1}$이 주어졌을 때, $p_t$를 알 수 있다.

### 1.2 Marginal Rate of Substitution/Stochastic Discount Factor

the basic consumption-based pricing equation은 다음과 같이 나타낼 수 있다.  

### $$p = E(mx)$$
### $$m = \beta \frac {u'(c_{t+1})} {u'(c_t)}$$  

$m_{t+1}$은 stochastic discount factor

불확실성이 없는 경우 현재 가치 공식으로 다음과 같이 표현 가능하다.

### $$p_t = \frac {1} {R^f} x_{t+1}$$

$R^f$는 gross risk-free rate

각 위험자산 i별로 할인하면 (자산별 위험 조정 할인)

### $$p_t^i = \frac {1} {R^i} E_t(x_{t+1}^i)$$

$m_{t+1}$은 marginal rate of substitution, pricing kernel이라고도 쓴다.

## Chapter 2. Applying the Basic Model

### 2.1 Assumptions and Applicability

위 공식들은 아래의 가정을 가정하지 않았다. (아직 필요하지 않음)  
　  
　  
**1. 시장이 완성되었거나 대표투자자가 있다.**
- 다른 투자자나 다른 자산의 존재 여부와 무관하게 각 개별 투자자가 자산에 접근 가능  

**2. 자산 수익 또는 보상은 정상적으로 분배되거나(옵션 없음), 시간이 지남에 따라 독립적이다.**
- 수익률이 정규분포이거나 Utility가 이차적이라고 가정하지 않음.

**3. Two-period investors, quadratic utility, or separable utility**
- 시간에 따른 I.I.D. 수익률을 가정하지 않음.

**4. 투자자는 인적 자본이나 노동 소득이 없다.**
- 시장화할 수 없는 외부 수입원 존재 가능성

**5. 시장이 평형에 도달했거나, 개인이 원하는 모든 유가증권을 사들였다.**
- 투자자가 원하는 자산을 모두 매수했다고 가정하지 않음.
