## 7장. 자본자산 가격결정 모형


### 7.1 시장균형

모든 투자자가 동일한 추정 모수로 평균-분산 모형을 사용하는 상황에서  
유일 펀드 = 시장 포트폴리오 = 효율적 펀드 = 최적 포트폴리오  
균형 이론: 자산 수익은 초기 가격과 최종 가격으로 결정 -> 투자자는 동일한 추정 값을 사용해 평균-분산 포트폴리오 문제 해결  
　　　　　-> 포트폴리오 주문 -> 가격 결정 -> 수요와 공급이 일치될때까지 반복


### 7.2 자본시장선 (CML)

### $\bar r = r_f + \frac {\bar r_M - r_f} {\sigma_M}$


### 7.3 가격결정 모형

### $\bar r_i - r_f = \beta_i(\bar r_M - r_f)$
### $\beta_i = \frac {\sigma_{iM}} {\sigma_M^2}$
 
보통주 베타: 공격적 기업은 베타가 높고, 보수적 기업은 베타가 낮음. 비슷한 사업하면 베타도 비슷.  
포트폴리오 베타: 포트폴리오를 구상하는 각 자산의 가중평균 합  


### 7.4 증권시장선

CAPM은 공분산, 베타와 선형적이다.  
$r_i = r_f + \beta_i(\bar r_M - r_f) + \epsilon_i$ 이기 때문에 $\sigma_i^2 = \beta_i^2\sigma_M^2 + var(\epsilon_i)$  
즉, 자산의 리스크 = 시장 리스크 + 시장과 관계 없는 개별 리스크  
시장과 관계 없는 개별 리스크를 비체계적 위험(nonsystematic, idiosyncratic, specific risk)이라 하고, 분산을 통해 줄일 수 있다.  


### 7.5 투자 영향

CAPM을 통해 Markowitz 문제에 대한 해가 위험 자산으로 구성된 유일펀드는 시장 포트폴리오라는 사실을 알 수 있다.  
CAPM에 동의하는 투자자는 전체 시장 비중에서 자산이 차지하는 상대적 비중만큼 자산을 구입해야한다.  
CAPM은 모든 투자자가 동일한 정보를 가지고 있다고 가정하지만 누군가 우수한 정보를 가지고 있다면, 시장을 뛰어넘는 포트폴리오를 구성할 수 있다.  


### 7.6 성과 평가

J를 토대로 펀드의 성과를 측정  

단계 1. 평균 수익률, 표준편차, 기하 평균 계산  
단계 2. 시장 포트폴리오와 무위험 자산 수익률 추출 -> 베타 계산  
단계 3. Jensen 척도: 기대 수익률을 단계 1의 평균 수익률로 대체. 즉, J를 알파로 인식 (cf. sharpe는 기울기를, J는 높이를 결정)   
단계 4. J가 높다고 효율적이라고 말할 수는 없다. 자본시장선 안에 존재해야함. 샤프지수(기울기)로 비교  


### 7.7 CAPM 가격 결정 모형

CAPM 가격 결정

### $P = \frac {\bar Q} {1 + r_f + \beta_i(\bar r_M - r_f)}$

확정등가 가격 결정

### $P = \frac {1} {1 + r_f} [\bar Q - \frac {cov(Q,r_M)(\bar r_M - r_f))} {\sigma_M^2}]$

P는 매입가격, Q는 매도가격

Tangency Portfolio는 샤프 비율이 가장 높은 포트폴리오이다. 모든 투자자가 평균-분산 최적화 모형을 사용하고, 자산의 확률 특성에 대해 모두 동의할 경우, 자산 수익에 대한 동일한 평균, 동일한 분산, 동일한 공분산을 사용하게 된다. 모든 투자자는 동일한 모수를 사용하기 때문에 동일한 위험 자산을 구성하고, 모두가 유일 펀드만 구입한다. 이 펀드가 시장을 형성하게 되면 유일 펀드는 시장 포트폴리오가 된다.  

CAL 식은 $E(R_p) = R_f + \frac {E(R_t) - R_f} {\sigma_t} * \sigma_p$  
CML 식은 $E(R_p) = R_f + \frac {E(R_m) - R_f} {\sigma_m} * \sigma_p$

Tangency Portfolio는 CAPM에서 시장 포트폴리오와 동일하므로  $R_m = R_t$, $\sigma_m = \sigma_t$ 가 되어 CAL 식과 CML 식은 같아진다.

임의 $\alpha$에 대해, 자산 $i$의 투자 비중은 $\alpha$이고 시장 포트폴리오 $M$의 투자 비중은 $1 - \alpha$인 포트폴리오를 생각해보자.  
포트폴리오의 기대수익률과 표준편차는 다음과 같다.  

$\bar r_\alpha = \alpha \bar r_i + (1-\alpha)\bar r_M$  
$\sigma_\alpha = [\alpha^2 \sigma_i^2 + 2\alpha (1-\alpha)\sigma_{iM} + (1-\alpha)^2\sigma^2_M]^{1/2}$  

$\alpha$값이 변할 때, 평균-분산 그래프에 곡선으로 나타나며, $\alpha = 0$이면 시장 포트폴리오 M에 해당한다. 따라서 $\alpha$가 0을 지날 때, 곡선은 점 M에서 자본시장선에 접한다. 점 M에서 자본시장선의 기울기와 곡선의 기울기가 같다. 이를 이용해 $\alpha$에 대해 두 식을 미분하면 아래와 같다.  

### $\frac {d\bar r_\alpha} {d\alpha} = \bar r_i - \bar r_M$  

### $\frac {d\bar \sigma_\alpha} {d\alpha} = \frac {\alpha \sigma_i^2 + (1-2\alpha)\sigma_{iM} + (\alpha-1)\sigma^2_M} {\sigma_\alpha}$

$\alpha = 0$을 대입해 계산하면 다음과 같다.

### $\frac {d\bar r_\alpha} {d\sigma_\alpha} = \frac {d\bar r_\alpha /  d\alpha} {d\bar \sigma_\alpha /  d\alpha}$  
### $\frac {d\sigma_\alpha} {d\alpha} = \frac {\sigma_{iM} - \sigma_M^2} {\sigma_M}$  
### $\frac {d\bar r_\alpha} {d\sigma_\alpha} = \frac {(\bar r_i - \bar r_M)\sigma_M} {\sigma_{iM} - \sigma_M^2}$  

점 M에서 곡선의 기울기는 자본시장선의 기울기와 같다.

### $\frac {(\bar r_i - \bar r_M)\sigma_M} {\sigma_{iM} - \sigma_M^2} = \frac {\bar r_M - r_f} {\sigma_M}$

$\bar r_i$에 대해 위 식을 정리하면 다음과 같다.

### $\bar r_i = r_f + (\frac {\bar r_M - r_f} {\sigma_M^2})\sigma_{iM} = r_f + \beta_i(\bar r_M - r_f)$
