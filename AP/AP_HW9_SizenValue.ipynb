### INVESTABLE

### INTUITIVE

많은 학자들은 가치 프리미엄이 실제로는 EMH와 모순되는 예외적인 현상이며 투자자들이 평가를 잘못한 결과라고 한다.  
행동주의 학파: 
기업을 평가할 때, 단순 성장률로 판단하면 성장기업은 지속적으로 고평가되며, 가치 기업은 지속적으로 저평가된다.  
투자자들은 인기 성장주에 익숙한 경향이 있다.  

Risk-based
- 가치주에 존재하는 부실 요인 척도: 배당금 25% 이상 삭감, 높은 자본 대비 부채 비율, 높은 수익 표준편차  
- 가치주는 레버리지가 높고 수익성 리스크가 큰 부실 기업이기 때문에 가격이 싸다.
- 이러한 위험 때문에 더 높은 수익을 제공한다.  
- 가치주는 경기가 나쁠 때는 성장주보다 훨씬 위험하고, 경기가 좋을 때는 성장주보다 덜 위험하다.  
- 가치주는 일반적으로 비생산적인 자본을 가진 기업이기 때문이다.  
- 생산 시설은 한번 구추고디면 줄이기 어렵다: 가치주는 성장주보다 훨씬 많이 보유한다.  
- 소비의 한계효용이 가장 높은 경기 침체기에 가치주가 낮은 수익률을 보인다. (가치주는 성장주보다 경기 순환에 민감)  

이러한 사실이 가치 프리미엄을 만든다.  

성장주: 듀레이션이 긴 자산, 장기 채권과 유사
가치주: 듀레이션이 낮은 자산, 단기 채권과 유사

성장주에 비해 실적, 수익성이 낮은 가치주는 디폴트 위험에 더 취약하며, 레버리지가 증가함에 따라 투자자들은 가치주에 더 높은 수익을 요구한다.  

가치 프리미엄과 경제
- 경제 호황에는 산업 생산이 증가하면 가치주가 성장주에 비해 상대적 위험도가 낮아진다. 따라서 가치주는 성장주보다 더 고평가된다. 그 결과, 시가총액이 높은 종목과 낮은 종목의 스프레드가 좁혀지고 가치 프리미엄이 하락한다.  
- 가치 프리미엄과 통화 공급량에는 비슷한 음의 관계가 존재한다. 통화 공급이 증가하면 주가가 상승하고, 가치주 가격은 성장주보다 더 많이 상승하는 경향이 있으며, 가치 프리미엄이 축소된다.  
- 가치 프리미엄과 금리 사이에는 양의 관계가 존재한다. 장기 금리가 상승하면 주식은 채권보다 매력이 떨어지고 주가는 하락하며, 가치주 가격이 성장주보다 빠르게 하락한다. 이는 가치 프리미엄의 증가로 이어진다.
- 가치주는 악재에 민감하고, 성장주는 호재에 민감하다. 
- 즉, 가치 프리미엄은 거시경제 위험을 통해 발생한다.  

가치주는 시장보다 변동성이 더 컸다.
- 1927년부터 2015년까지 대형주 인덱스 연간 표준편차는 19.7%, 대형 가치주 인덱스는 26.8%


### BEHAVIORAL EXPLANATIONS (MISPRICINGS)

투자자들은 성장주 실적에 대한 기대치를 지나치게 낙관적으로 보고, 가치주 실적은 지나치게 비관적으로 본다.

### 1-2. Z-Scoring

```python
# 결측치 제거
value_col = ['PBR','PER','PSR','DIVY']
value[value_col] = value[value_col].fillna(0)
value.isnull().sum()
```

```python
# 결측치 제거
size['Size'].fillna(0, inplace = True)
size.isnull().sum()
```

### 1-3. Score Aggregation

```python
value['value'] = value[['PBR','PER','PSR']].sum(axis=1) - value['DIVY']
vns_df = value[['cusip','mdate','cap','return','value']]
vns_df['size'] = size['Size']
vns_df
```

### 1-4. Construct Long Portfolio

```python
# 매수종목 선정
vns_df['value_cond'] = vns_df.groupby('mdate')['value'].transform(lambda x: x < x.quantile(0.3))
vns_df['size_cond'] = vns_df.groupby('mdate')['size'].transform(lambda x: x < x.quantile(0.3))

vns_df.set_index('cusip', inplace = True)
```

```python
# 기간 설정
month = vns_df['mdate'].drop_duplicates().values
period = 18
```

```python
# Value 백테스팅
# 예시: 1월 1일 시가로 매수 -> 다음년 6월 30일 종가로 매도 -> 7월 1일 시가로 매수 -> 다음년 12월 31일 종가로 매도
rtn = {}

for m in month:  # 한달 간격 수익률 체크
    # 18개월마다 리밸런싱
    if period == 18:
        buy_stocks_df = vns_df[(vns_df['mdate']==m) & (vns_df['value_cond']==True) & (vns_df['return'].notna())]  # 종목 선정
        buy_stocks = buy_stocks_df.index  # 종목 설정
        weights = buy_stocks_df['cap'] / buy_stocks_df['cap'].sum()  # 가중치 설정
        buy_set = pd.Series(data = weights.values, index = buy_stocks, name = 'weight')
        period = 0
        print(f'=====  {m[:-2]}01 리밸런싱  =====')
    period += 1
    
    # 수익률 확인
    monthly_df = pd.merge(buy_set, vns_df[vns_df['mdate']==m]['return'], how='left', on='cusip')
    
    # 매수종목 사라지면 상폐(수익률: -1)로 판정하고 매수목록에서 제거
    if monthly_df['return'].isnull().sum() > 0: 
        drop_index = monthly_df[monthly_df['return'].isnull()].index  # 사라진 종목 체크
        buy_set.drop(index = drop_index, inplace = True)  # 매수목록에서 제거
        monthly_df.loc[drop_index,'return'] = -1 # delist return을 위한 수익률 처리
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 추가
        print(f'- {m[:-3]} : 상장폐지 {len(drop_index)} 종목')
    else:
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()
        
value_rtn = pd.Series(rtn)
```

```python
# 기간 설정
month = vns_df['mdate'].drop_duplicates().values
period = 18
```

```python
# Size 백테스팅
# 예시: 1월 1일 시가로 매수 -> 다음년 6월 30일 종가로 매도 -> 7월 1일 시가로 매수 -> 다음년 12월 31일 종가로 매도
rtn = {}

for m in month:  # 한달 간격 수익률 체크
    # 18개월마다 리밸런싱
    if period == 18:
        buy_stocks_df = vns_df[(vns_df['mdate']==m) & (vns_df['size_cond']==True) & (vns_df['return'].notna())]  # 종목 선정
        buy_stocks = buy_stocks_df.index  # 종목 설정
        weights = buy_stocks_df['cap'] / buy_stocks_df['cap'].sum()  # 가중치 설정
        buy_set = pd.Series(data = weights.values, index = buy_stocks, name = 'weight')
        period = 0
        print(f'=====  {m[:-2]}01 리밸런싱  =====')
    period += 1
    
    # 수익률 확인
    monthly_df = pd.merge(buy_set, vns_df[vns_df['mdate']==m]['return'], how='left', on='cusip')
    
    # 매수종목 사라지면 상폐(수익률: -1)로 판정하고 매수목록에서 제거
    if monthly_df['return'].isnull().sum() > 0: 
        drop_index = monthly_df[monthly_df['return'].isnull()].index  # 사라진 종목 체크
        buy_set.drop(index = drop_index, inplace = True)  # 매수목록에서 제거
        monthly_df.loc[drop_index,'return'] = -1 # delist return을 위한 수익률 처리
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 추가
        print(f'- {m[:-3]} : 상장폐지 {len(drop_index)} 종목')
    else:
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()
        
size_rtn = pd.Series(rtn)
```

### 1-5. Comparison to Benchmark

```python
def result_analysis(rtn:pd.Series, period, freq):
    rtn_cum = (rtn+1).cumprod()
    cagr = rtn_cum.iloc[-1]**(1/period) - 1
    rtn_std = np.std(rtn) * np.sqrt(freq)
    sharpe = cagr / rtn_std
    print(f'기간: {rtn.index[0]} ~ {rtn.index[-1]}')
    print(f'누적수익률: {rtn_cum.iloc[-1]:.2%}')
    print(f'평균 월 수익률: {rtn.mean():.2%}')
    print(f'CAGR: {cagr:.2%}')
    print(f'Sharpe Ratio: {sharpe:.2f}')
```

```python
# 시장포트폴리오
mkt = mkt.query("mdate <= '2021-12-31'")
mkt.index = mkt['mdate']
mkt_rtn = mkt['market'] - mkt['risk_free']

print('--- Market ---')
result_analysis(mkt_rtn, period = 52, freq = 12)
```

```python
# Value
print('--- High Value ---')
result_analysis(value_rtn, period = 52, freq = 12)
```

```python
# Size
print('--- Small Size ---')
result_analysis(size_rtn, period = 52, freq = 12)
```

### 1-6.  Visualization

```python
def result_visualization(df):
    df_cum = (df+1).cumprod()
    df_high = df_cum.cummax()
    df_mdd = df_cum / df_high - 1
    
    fig, ax = plt.subplots(2, figsize = (15,8), gridspec_kw={'height_ratios': [2.5, 1], 'hspace':0.05}, sharex = True)
    (df_cum - 1).plot(ax=ax[0], title = 'Cumulative Return', grid=True)    
    df_mdd.plot(ax=ax[1], grid=True)
```

```python
vns_rtn = pd.DataFrame([mkt_rtn, size_rtn, value_rtn], index = ['market','size','value']).T
result_visualization(vns_rtn)
```

```python
# 매수종목 선정
vns_ls_df = vns_df[['mdate','cap','return','value','size']].copy()
vns_ls_df['beta'] = size['beta'].values

vns_ls_df['value_long_cond'] = vns_ls_df.groupby('mdate')['value'].transform(lambda x: x < x.quantile(0.3))
vns_ls_df['value_short_cond'] = vns_ls_df.groupby('mdate')['value'].transform(lambda x: x > x.quantile(0.7))

vns_ls_df['size_long_cond'] = vns_ls_df.groupby('mdate')['size'].transform(lambda x: x < x.quantile(0.3))
vns_ls_df['size_short_cond'] = vns_ls_df.groupby('mdate')['size'].transform(lambda x: x > x.quantile(0.7))

vns_ls_df
```

```python
# 기간 설정
month = vns_ls_df['mdate'].drop_duplicates().values
period = 18
```

```python
# Value
rtn = {}

for m in month:
    # 18개월마다 리밸런싱
    if period == 18:
        buy_stocks_df = vns_ls_df[(vns_ls_df['mdate']==m) & 
                                  ((vns_ls_df['value_long_cond']==True)|(vns_ls_df['value_short_cond'])) & 
                                  (vns_ls_df['return'].notna())]  # 종목 선정
        long_weight = buy_stocks_df[buy_stocks_df['value_long_cond']==True]['beta'].sum() / buy_stocks_df['beta'].sum()
        short_weight = buy_stocks_df[buy_stocks_df['value_short_cond']==True]['beta'].sum() / buy_stocks_df['beta'].sum()
        buy_stocks_df['weight'] = 0
        buy_stocks_long_cap = buy_stocks_df['cap'][buy_stocks_df['value_long_cond']==True]
        buy_stocks_short_cap = buy_stocks_df['cap'][buy_stocks_df['value_short_cond']==True]
        buy_stocks_df['weight'][buy_stocks_df['value_long_cond']==True] = buy_stocks_long_cap/buy_stocks_long_cap.sum()*long_weight
        buy_stocks_df['weight'][buy_stocks_df['value_short_cond']==True] = -1*buy_stocks_short_cap/buy_stocks_short_cap.sum()*short_weight
        buy_set = pd.Series(buy_stocks_df['weight'])
        period = 0
        print(f'=====  {m[:-2]}01 리밸런싱  =====')
    period += 1
    
    # 수익률 확인
    monthly_df = pd.merge(buy_set, vns_ls_df[vns_ls_df['mdate']==m]['return'], how='left', on='cusip')
    
    
    # 매수종목 사라지면 상폐(수익률: -1)로 판정하고 매수목록에서 제거
    if monthly_df['return'].isnull().sum() > 0: 
        drop_index = monthly_df[monthly_df['return'].isnull()].index  # 사라진 종목 체크
        buy_set.drop(index = drop_index, inplace = True)  # 매수목록에서 제거
        monthly_df.loc[drop_index,'return'] = -1 # delist return을 위한 수익률 처리
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 추가
        print(f'- {m[:-3]} : 상장폐지 {len(drop_index)} 종목')
    else:
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()

value_ls_rtn = pd.Series(rtn)
```

```python
# 기간 설정
month = vns_ls_df['mdate'].drop_duplicates().values
period = 18
```

```python
# Size
rtn = {}

for m in month:
    # 18개월마다 리밸런싱
    if period == 18:
        buy_stocks_df = vns_ls_df[(vns_ls_df['mdate']==m) & 
                                  ((vns_ls_df['size_long_cond']==True)|(vns_ls_df['size_short_cond'])) & 
                                  (vns_ls_df['return'].notna())]  # 종목 선정
        long_weight = buy_stocks_df[buy_stocks_df['size_short_cond']==True]['beta'].sum() / buy_stocks_df['beta'].sum()
        short_weight = buy_stocks_df[buy_stocks_df['size_long_cond']==True]['beta'].sum() / buy_stocks_df['beta'].sum()
        buy_stocks_df['weight'] = 0
        buy_stocks_long_cap = buy_stocks_df['cap'][buy_stocks_df['size_long_cond']==True]
        buy_stocks_short_cap = buy_stocks_df['cap'][buy_stocks_df['size_short_cond']==True]
        buy_stocks_df['weight'][buy_stocks_df['size_long_cond']==True] = buy_stocks_long_cap/buy_stocks_long_cap.sum()*long_weight
        buy_stocks_df['weight'][buy_stocks_df['size_short_cond']==True] = -1*buy_stocks_short_cap/buy_stocks_short_cap.sum()*short_weight
        buy_set = pd.Series(buy_stocks_df['weight'])
        period = 0
        print(f'=====  {m[:-2]}01 리밸런싱  =====')
    period += 1
    
    # 수익률 확인
    monthly_df = pd.merge(buy_set, vns_ls_df[vns_ls_df['mdate']==m]['return'], how='left', on='cusip')
    
    
    # 매수종목 사라지면 상폐(수익률: -1)로 판정하고 매수목록에서 제거
    if monthly_df['return'].isnull().sum() > 0: 
        drop_index = monthly_df[monthly_df['return'].isnull()].index  # 사라진 종목 체크
        buy_set.drop(index = drop_index, inplace = True)  # 매수목록에서 제거
        monthly_df.loc[drop_index,'return'] = -1 # delist return을 위한 수익률 처리
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 추가
        print(f'- {m[:-3]} : 상장폐지 {len(drop_index)} 종목')
    else:
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()

size_ls_rtn = pd.Series(rtn)
```
