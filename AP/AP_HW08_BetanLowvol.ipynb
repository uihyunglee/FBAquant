### EXPLAINING THE ANOMALY

CAPM의 레버리지나 공매도 제약이 없다는 가정, 거래 비용이나 세금이 없다는 가정은 현실을 반영하지 못한다.  
* ex) 공매도 비용이 높은 주식은 잘못 평가  
* 이 때문에 수익률을 올리기 위해 리스크 프리미엄이 더 큰, 즉 베타가 높은 종목에 대해 수요가 높아짐.
* 따라서 하이 리스크 종목은 소유자의 편향성이 크기 때문에 로우 리스크 종목보다 가격이 비싸게 형성될 가능성이 높음.  

CAPM은 투자자가 위험을 회피하고, 기대 효용을 극대화하며 수익률의 평균과 분산에 따라 투자한다고 가정하지만, 현실은 그렇지 않다.  
* ex) IPO, 페니주, 파산주

매니저가 높은 수익을 얻기 위해 변동성이 큰 포트폴리오를 추구한다. 

정보의 비대칭이 존재한다.

## Problem 1. Low Volatility – Long Portfolio

### 1-1. Data Load

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings( 'ignore' )

plt.rcParams['font.family'] = ['Malgun Gothic', 'AppleGothic']
plt.rcParams['axes.unicode_minus'] = False

data_path = 'data/'

mkt_js = pd.read_json(data_path + 'Market.json')
lv_js = pd.read_json(data_path + 'LowVolatility.json')

mkt = pd.DataFrame(dict(mkt_js.iloc[:,0])).T
lv = pd.DataFrame(dict(lv_js.iloc[:,0])).T
```

```python
mkt
```

```python
lv
```

### 1-2. Z-Scoring

```python
target_col = ['5yBeta','5yROEVol','5yPriceVol']
z_col = ['Z_'+c for c in target_col]

scale_df = lv[['mdate']+target_col].copy()

lv[z_col] = ((scale_df[target_col] - scale_df.groupby('mdate').transform(np.mean)) / scale_df.groupby('mdate').transform(np.std))
lv[z_col] = lv[z_col].fillna(0)
lv[z_col]
```

```python
print('**NaN count**')
print(lv[z_col].isnull().sum())
print('\n**Mean**')
print(lv[z_col].mean())
print('\n**Standard Deviation**')
print(lv[z_col].std())
```

### 1-3. Score Aggregation

```python
lv['score'] = lv[z_col].sum(axis=1)
lvs = lv[['cusip','mdate','cap','return','score']]
lvs
```

### 1-4. Construct Long Portfolio.

```python
# 매수종목 선정
lvs['cond'] = lvs.groupby('mdate')['score'].transform(lambda x: x < x.quantile(0.3))
lvs.set_index('cusip', inplace = True)
```

```python
# 기간 설정
month = lvs['mdate'].drop_duplicates().values
period = 18
```

```python
# 백테스팅
# 예시: 1월 1일 시가로 매수 -> 다음년 6월 30일 종가로 매도 -> 7월 1일 시가로 매수 -> 다음년 12월 31일 종가로 매도
rtn = {}
rtn_delist = {}

for m in month:  # 한달 간격 수익률 체크
    # 18개월마다 리밸런싱
    if period == 18:
        buy_stocks_df = lvs[(lvs['mdate']==m) & (lvs['cond']==True) & (lvs['return'].notna())]  # 종목 선정
        buy_stocks = buy_stocks_df.index  # 종목 설정
        weights = buy_stocks_df['cap'] / buy_stocks_df['cap'].sum()  # 가중치 설정
        buy_set = pd.Series(data = weights.values, index = buy_stocks, name = 'weight')
        period = 0
        print(f'=====  {m[:-2]}01 리밸런싱  =====')
    period += 1
    
    # 수익률 확인
    monthly_df = pd.merge(buy_set, lvs[lvs['mdate']==m]['return'], how='left', on='cusip')
    
    
    # 매수종목 사라지면 상폐(수익률: -1)로 판정하고 매수목록에서 제거
    if monthly_df['return'].isnull().sum() > 0: 
        drop_index = monthly_df[monthly_df['return'].isnull()].index  # 사라진 종목 체크
        buy_set.drop(index = drop_index, inplace = True)  # 매수목록에서 제거
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 추가
        monthly_df.loc[drop_index,'return'] = -1  # delist return을 위한 수익률 처리
        print(f'- {m[:-3]} : 상장폐지 {len(drop_index)} 종목')
    else:
        rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()
    
    rtn_delist[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 처리
    
lv_rtn = pd.Series(rtn)
lv_rtn_delist = pd.Series(rtn_delist)
```

### 1-5. Comparison to Benchmark

```python
# s&p500
mkt = mkt.query("mdate <= '2021-12-31'")
snp_excess_rtn = mkt['SnP500'] - mkt['risk_free']
snp_excess_cum = (snp_excess_rtn+1).cumprod()
snp_excess_cgar = snp_excess_cum.iloc[-1]**(1/52) - 1
snp_std = np.std(snp_excess_rtn) * np.sqrt(12)
snp_sharpe = snp_excess_cgar / snp_std
```

```python
# Low Vol
lv_excess_rtn = lv_rtn - mkt['risk_free'].values
lv_excess_cum = (lv_excess_rtn+1).cumprod()
lv_excess_cgar = lv_excess_cum.iloc[-1]**(1/52) - 1
lv_std = np.std(lv_excess_rtn) * np.sqrt(12)
lv_sharpe = lv_excess_cgar / lv_std
```

```python
# Low Vol 상장폐지 수익률 -1로 넣었을 때
lv_delist_excess_rtn = lv_rtn_delist - mkt['risk_free'].values
lv_delist_excess_cum = (lv_delist_excess_rtn+1).cumprod()
lv_delist_excess_cgar = lv_delist_excess_cum.iloc[-1]**(1/52) - 1
lv_delist_std = np.std(lv_delist_excess_rtn) * np.sqrt(12)
lv_delist_sharpe = lv_delist_excess_cgar / lv_delist_std
```

```python
# Information Ratio
lv_snp_rtn = lv_rtn - mkt['SnP500'].values
lv_snp_cum = (lv_snp_rtn+1).cumprod()
lv_snp_cgar = lv_snp_cum.iloc[-1]**(1/52) - 1
lv_snp_std = np.std(lv_snp_rtn) * np.sqrt(12)
lv_snp_info = lv_snp_cgar / lv_snp_std
```

```python
print(f'Low Vol Portfolio Sharpe : {lv_sharpe:.4f}')
print(f'S&P500 Sharpe : {snp_sharpe:.4f}')
print(f'Low Vol - S&P500 Information Ratio : {lv_snp_info:.4f}')
```

```python
print(f'Low Vol Portfolio Sharpe (상폐종목 수익률 -1 적용) : {lv_delist_sharpe:.4f}')
```

### 1-6.  Visualization

```python
lv_cum = (lv_rtn+1).cumprod()
snp_cum = (mkt['SnP500']+1).cumprod()
lv_delist_cum = (lv_rtn_delist+1).cumprod()

lv_high = lv_cum.cummax()
snp_high = snp_cum.cummax()
lv_delist_high = lv_delist_cum.cummax()

lv_mdd = lv_cum / lv_high
snp_mdd = snp_cum / snp_high
lv_delist_mdd = lv_delist_cum / lv_delist_high
```

```python
lv_cum.plot(figsize=(14,6), title='Low Vol - S&P500 Return', label='Low Vol', legend=True, color = 'black')
snp_cum.plot(label='S&p500', legend=True);
lv_delist_cum.plot(label='Low Vol (delist)', legend=True,);
```

```python
lv_mdd.plot(figsize=(14,4), title='Low Vol - S&P500 DrawDown', label='Low Vol', legend=True, color = 'black')
snp_mdd.plot(label='S&p500', legend=True);
lv_delist_mdd.plot(label='Low Vol (delist)', legend=True,);
plt.grid(True)
```

## Problem 2. Low Volatility – Long/Short Portfolio

### 2-1 ~ 2-3 위와 동일

```python
lvs = lv[['cusip','mdate','cap','return','score','beta']]
lvs
```

### 2-4. Construct Long/Short Portfolio

```python
# 매수종목 선정
lvs['long_cond'] = lvs.groupby('mdate')['score'].transform(lambda x: x < x.quantile(0.3))
lvs['short_cond'] = lvs.groupby('mdate')['score'].transform(lambda x: x > x.quantile(0.7))
lvs.set_index('cusip', inplace = True)
lvs
```

```python
# 기간 설정
month = lvs['mdate'].drop_duplicates().values
period = 18
```

```python
# 백테스팅
# 예시: 1월 1일 시가로 매수 -> 다음년 6월 30일 종가로 매도 -> 7월 1일 시가로 매수 -> 다음년 12월 31일 종가로 매도
ls_rtn = {}
ls_rtn_delist = {}

for m in month:  # 한달 간격 수익률 체크
    # 18개월마다 리밸런싱
    if period == 18:
        buy_stocks_df = lvs[(lvs['mdate']==m) & ((lvs['long_cond']==True)|(lvs['short_cond'])) & (lvs['return'].notna())]  # 종목 선정
        long_weight = buy_stocks_df[buy_stocks_df['short_cond']==True]['beta'].sum() / buy_stocks_df['beta'].sum()
        short_weight = buy_stocks_df[buy_stocks_df['long_cond']==True]['beta'].sum() / buy_stocks_df['beta'].sum()
        buy_stocks_df['weight'] = 0
        buy_stocks_df['weight'][buy_stocks_df['long_cond']==True] = buy_stocks_df['cap'][buy_stocks_df['long_cond']==True]/buy_stocks_df['cap'][buy_stocks_df['long_cond']==True].sum()*long_weight
        buy_stocks_df['weight'][buy_stocks_df['short_cond']==True] = -1*buy_stocks_df['cap'][buy_stocks_df['short_cond']==True]/buy_stocks_df['cap'][buy_stocks_df['short_cond']==True].sum()*short_weight
        buy_set = pd.Series(buy_stocks_df['weight'])
        period = 0
        print(f'=====  {m[:-2]}01 리밸런싱  =====')
    period += 1
    
    # 수익률 확인
    monthly_df = pd.merge(buy_set, lvs[lvs['mdate']==m]['return'], how='left', on='cusip')
    
    
    # 매수종목 사라지면 상폐(수익률: -1)로 판정하고 매수목록에서 제거
    if monthly_df['return'].isnull().sum() > 0: 
        drop_index = monthly_df[monthly_df['return'].isnull()].index  # 사라진 종목 체크
        buy_set.drop(index = drop_index, inplace = True)  # 매수목록에서 제거
        ls_rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 추가
        monthly_df.loc[drop_index,'return'] = -1  # delist return을 위한 수익률 처리
        print(f'- {m[:-3]} : 상장폐지 {len(drop_index)} 종목')
    else:
        ls_rtn[m] = (monthly_df['weight'] * monthly_df['return']).sum()
    
    ls_rtn_delist[m] = (monthly_df['weight'] * monthly_df['return']).sum()  # 수익률 처리
    
lv_ls_rtn = pd.Series(ls_rtn)
lv_ls_rtn_delist = pd.Series(ls_rtn_delist)
```

### 2-5. Statistical Robustness of Alpha 

```python
lv_ls_rtn
```

```python
from scipy import stats

stats.ttest_1samp(lv_ls_rtn, popmean=0, alternative='greater')
```

p-value가 0에 근사하므로 $\alpha$는 0보다 크다는 귀무가설을 채택한다.

### 2-6. Visualization

```python
lv_ls_cum = (lv_ls_rtn+1).cumprod()
lv_ls_delist_cum = (lv_ls_rtn_delist+1).cumprod()

lv_ls_high = lv_ls_cum.cummax()
lv_ls_delist_high = lv_ls_delist_cum.cummax()

lv_ls_mdd = lv_ls_cum / lv_ls_high
lv_ls_delist_mdd = lv_ls_delist_cum / lv_ls_delist_high
```
